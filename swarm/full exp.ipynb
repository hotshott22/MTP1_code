```python
# -*- coding: utf-8 -*-
"""swarm_with_grover.ipynb - Modified with Multiple Obstacles

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1zRTbBweWntX9ldaR885Moq7nv0_6me0D
"""

!pip install qiskit

!pip install qiskit-aer

# qiskit tools for quantum circuit construction, visualization and simulation
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit_aer import Aer # Import Aer from qiskit_aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# utilities
from operator import attrgetter, itemgetter
from scipy.spatial.distance import cityblock
from heapq import nlargest
import numpy as np

class Target:
    def __init__(self, name, x, y): # no indetermination in the target's position
        self.name = name
        self.x = x
        self.y = y

T = Target("T", 0.875, 0.125) # deep in the ocean

def reward(T, betax, betay):
    return 1 - (((T.x - betax)**2 + (T.y - betay)**2)**0.5)/np.sqrt(2) # distance is normalized on sqrt(2)

class Obstacle:
    _registry = []

    def __init__(self, name, x, y):
        self._registry.append(self)
        self.name = name
        self.x = x
        self.y = y

# Create multiple obstacles at different locations around the target
O1 = Obstacle("O1", 0.125, 0.125)  # Original obstacle - bottom left
O2 = Obstacle("O2", 0.875, 0.375)  # Near target, different cell
O3 = Obstacle("O3", 0.625, 0.125)  # Between start and target
O4 = Obstacle("O4", 0.375, 0.625)  # Mid-field obstacle
O5 = Obstacle("O5", 0.625, 0.875)  # Upper right area

# utility function that returns the complementary amplitude
# of ampl, evaluated as sqrt(1 - ||ampl||^2)
#
def c_ampl(ampl):
    return np.sqrt(1 - ampl**2)

class Robotx(object):
    _registry = []

    def __init__(self, name, alphax, alphay):
        self._registry.append(self)
        self.name = name
        self.alphax = alphax
        self.betax = c_ampl(alphax)
        self.alphay = alphay
        self.betay = c_ampl(alphay)
        self.delta = reward(T, self.betax, self.betay)
        self.gamma = c_ampl(self.delta)

# arbitrary number of robots that, at the start, are uniformly distributed
# in the square of side length radius*2, centered in starting_cluster_coord
#
num_of_robots = 10

# must be lower than 0.25, otherwise, the robots would start on different cells
radius = 0.05

# center of the (0,0) cell
starting_cluster_coord = (0.125, 0.875)

# boundaries to compute the initial random scattering on
a_x, a_y = c_ampl(starting_cluster_coord[0]-radius), c_ampl(starting_cluster_coord[0]+radius)
b_x, b_y = c_ampl(starting_cluster_coord[1]-radius), c_ampl(starting_cluster_coord[1]+radius)

# robots' initialization
for i in range(num_of_robots):
    Robotx('R'+str(i), np.random.uniform(a_x,a_y), np.random.uniform(b_x,b_y))

# note: values are stored with full precision, rounding is done only on visualization (with :.2f)
for k in Robotx._registry:
    print(f"{k.name} {k.betax:.2f} {k.betay:.2f} {k.gamma:.2f} {k.delta:.2f}")

def calculate_avg_reward():
    rewards = [robot.delta for robot in Robotx._registry]
    return np.mean(rewards)

print(f"Average reward: {calculate_avg_reward()}")

# this utility function generates the visualization of the environment with each of the entitites;
# the map is maked with a squared lattice, so as to outline the micro-cell subdivision and the robots
# obstacles, and target are color coded in the followig way:
#   robots: green circle;
#   obstacles: red pentagon;
#   target: turquoise star.
#
def plot_scatterplot(*args):

    # to print grid on map
    ax = plt.figure().gca()
    ax.set_xticks(np.arange(0, 1, 0.25))
    ax.set_yticks(np.arange(0, 1, 0.25))
    plt.grid()

    for i in Robotx._registry:
        # color gradient to better visualize the different robots within the map;
        # the higher the i-th index of the robot's name, the darker the green gradient.
        robot_color = hex(int("0x30ca15", 16) + int("0x00bb00", 16)*int(i.name[1])).replace("0x","#")
        plt.scatter(i.betax, i.betay, s = 300, marker = 'o', color = robot_color)

    plt.scatter(T.x, T.y, s = 400, marker = '*', color = 'turquoise')

    for o in Obstacle._registry:
        plt.scatter(o.x, o.y, s = 400, marker = 'p', color = 'red')

    plt.axis([0, 1, 0, 1])

    if args:
        plt.title(args[0])

    plt.show()

plot_scatterplot("Initialization with Multiple Obstacles")

for r in Robotx._registry:
    if (r.delta < 0.5):
        print(f"{r.name} {r.delta:.2f} achtung!") # and start from this point to build the reshuffle section

# returns a dict so that:
#  - the keys are the coordinates of the centroids of a square dim_map x dim_map map of side length 1
#  - the values are in the form (micro_cell, macro_cell) of the centroids
def find_centroids(dim_map):

    unit_displacement = 1 / (dim_map*2)

    centroids_coords = []
    for i in range(dim_map,0,-1):
        row = []
        for j in range(dim_map):
            row.append((j*unit_displacement*2 + unit_displacement, i*unit_displacement*2 - unit_displacement))
        centroids_coords.append(row)

    centroids_dict = {t: (f"{str(i%2)}{str(j%2)}", f"{0 if t[1] > 0.5 else 1}{0 if t[0] < 0.5 else 1}")
     for i,row in enumerate(centroids_coords) for j,t in enumerate(row)}

    return centroids_dict

centroids = find_centroids(4)
print("Centroids:\n{ " + "\n  ".join(f"{k}: {v} " for k, v in centroids.items()) + "}")

# returns the centroid tuple of the cell that a point of coordinates (x,y) is located in
def find_closest_centroid(centroids, x, y):
    centroids_coords = np.array(list(centroids.keys()))
    distances = np.linalg.norm(centroids_coords-np.array([x, y]), axis=1)
    return centroids[tuple(centroids_coords[np.argmin(distances)])]

for r in Robotx._registry:
    print(f"Closest centroid to {r.name}: {find_closest_centroid(centroids, r.betax, r.betay)}")

# associate centroids to obstacles
obstacles_centroids = {coord: v for (coord,v) in centroids.items() for o in Obstacle._registry if coord == (o.x, o.y)}
print("Obstacles' centroids:\n{ " + "\n  ".join(f"{k}: {v} " for k, v in obstacles_centroids.items()) + "}")

# takes as input strings with the binary encoding of cells and outputs the manhattan distance between them
def manhattan(start_cell, target_cell):
    start_cell_v, target_cell_v = [int(b) for b in start_cell], [int(b) for b in target_cell]
    return int(cityblock(start_cell_v, target_cell_v).item())

class GroverPlanner:
    def __init__(self, obstacles):
        self.simulator = Aer.get_backend("aer_simulator")
        self.obstacles = obstacles

    #
    # Utility functions
    #
    def XOR(self, qc, a, b, output):
        qc.cx(a, output)
        qc.cx(b, output)

    def encode_cell(self, qc, cell, qubits):
        for idx, bit in enumerate(cell):
            bitn = int(bit)
            if bitn:
                qc.x(qubits[idx])

    def encode_target(self, qc, target_cell, qubits):
        for idx, bit in enumerate(target_cell):
            bitn = int(bit)
            if not bitn:
                qc.x(qubits[idx])

    def obstacles_in_map(self, target_cell):
        return any(target_cell[1] == macro_cell for macro_cell in list(map(lambda ob_t: ob_t[1], self.obstacles)))

    #
    # Grover diffuser
    #
    def diffuser(self, nqubits):
        diff = QuantumCircuit(nqubits)

        # Apply transformation |s> -> |00..0> (H-gates)
        for qubit in range(nqubits):
            diff.h(qubit)

        # Apply transformation |00..0> -> |11..1> (X-gates)
        for qubit in range(nqubits):
            diff.x(qubit)

        # Do multi-controlled-Z gate
        diff.h(nqubits - 1)
        # Use mcx with mode='noancilla' for multi-controlled Toffoli
        diff.mcx(list(range(nqubits - 1)), nqubits - 1, mode='noancilla')
        diff.h(nqubits - 1)

        # Apply transformation |11..1> -> |00..0>
        for qubit in range(nqubits):
            diff.x(qubit)

        # Apply transformation |00..0> -> |s>
        for qubit in range(nqubits):
            diff.h(qubit)

        U_s = diff.to_gate()
        U_s.name = "U$_\psi$"
        return U_s

    #
    # Half oracle
    #
    # it performs the quantum moves to get to the final state the solution test is performed on; its structure
    # changes according to both the number of movements required to get to the target and the presence of obstacles
    #
    def half_oracle(self, num_of_M, obstacles_in_map, quantum_registers):

        if num_of_M == 1:

            ho = QuantumCircuit(*quantum_registers)
            pos_qubits, mov_qubits, c, output = quantum_registers

            # -- First M --
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])

        elif num_of_M == 2:

            ho = QuantumCircuit(*quantum_registers)

            if obstacles_in_map:
                pos_qubits, mov_qubits, c, o, caux, out, output = quantum_registers
            else:
                pos_qubits, mov_qubits, c, output = quantum_registers

            # -- First M --
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])

            if obstacles_in_map:
                # -- Obstacle encoding --
                for obstacle in self.obstacles:
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])
                    ho.ccx(c[0], c[1], o)
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])

            # -- Second M --
            self.XOR(ho, c[1], mov_qubits[1], c[2])
            ho.x(c[2])
            self.XOR(ho, c[0], mov_qubits[1], c[3])

            if obstacles_in_map:
                # -- Selection stage --
                ho.ccx(c[2], o, caux[0])
                ho.x(caux[0])
                ho.ccx(c[3], o, caux[1])
                ho.x(caux[1])

                ho.x(o)

                ho.ccx(pos_qubits[0], o, caux[2])
                ho.x(caux[2])
                ho.ccx(pos_qubits[1], o, caux[3])
                ho.x(caux[3])

                ho.ccx(caux[0], caux[2], out[0])
                ho.x(out[0])
                ho.ccx(caux[1], caux[3], out[1])
                ho.x(out[1])

        HalfOracle = ho.to_gate()
        HalfOracle.name = "HO"
        return HalfOracle

    #
    # Grover oracle
    #
    # it is composed of a oracle half, the testing operator, and the mirror circuit of the first half
    #
    def oracle(self, num_of_M, target_cell, m_type, obstacles_in_map, qc, quantum_registers):

        halforacle = self.half_oracle(num_of_M, obstacles_in_map, quantum_registers)

        additional_regs = []

        if obstacles_in_map:
            pos_qubits, mov_qubits, c, *additional_regs, output = quantum_registers
        else:
            pos_qubits, mov_qubits, c, output = quantum_registers

        # -- First half of the oracle --
        qc.append(
            halforacle,
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + (
                (additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:])
                if additional_regs
                else []
            )
            + output[:],
        )

        # -- Test operator --
        test_qubits = c[-2:] if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map) else additional_regs[2]
        tcell_idx = 0 if m_type == "micro" else 1
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])
        qc.ccx(test_qubits[0], test_qubits[1], output)
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])

        # -- Mirror circuit of the first half of the oracle --
        qc.append(
            halforacle.inverse(),
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + (
                (additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:])
                if additional_regs
                else []
            )
            + output[:],
        )

    #
    # Path planning
    #
    # takes as input the tuples of centroids start_cell and target_cell, together with the movement type (m_type),
    # that can yield the values "micro" or "macro"
    #
    def find_path(self, start_cell, target_cell, m_type):

        # number of M operators is equal to the Manhattan distance between start and target
        cell_idx = 0 if m_type == "micro" else 1
        num_of_M = manhattan(start_cell[cell_idx], target_cell[cell_idx])

        # --- Initializations ---
        pos_qubits = QuantumRegister(2, name="r")
        mov_qubits = QuantumRegister(num_of_M, name="m")
        c = QuantumRegister(2 * num_of_M, name="c")

        o = QuantumRegister(1, name="o")
        caux = QuantumRegister(4, name="caux")
        out = QuantumRegister(2, name="out")

        output = QuantumRegister(1, name="q")
        cbits = ClassicalRegister(2 + num_of_M, name="cbits")

        # check if there are obstacles to be considered within the path
        obstacles_in_map = self.obstacles_in_map(target_cell) and m_type == "micro"

        # ensemble of quantum registers to build the QuantumCircuit with
        base_registers = [pos_qubits, mov_qubits, c, output, cbits]
        additional_registers = base_registers.copy()
        additional_registers[3:3] = [o, caux, out]
        quantum_registers = (
            base_registers
            if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map)
            else additional_registers
        )

        qc = QuantumCircuit(*quantum_registers)

        # initialize pos_qubits in uniform superposition
        qc.h(pos_qubits)

        # initialize mov_qubits in state |s>
        qc.h(mov_qubits)

        # initialize 'output0' in state |->
        qc.initialize([1, -1] / np.sqrt(2), output)

        # --- Oracle + Diffuser ---
        num_iterations = 1
        for _ in range(num_iterations):
            self.oracle(
                num_of_M,
                target_cell,
                m_type,
                obstacles_in_map,
                qc,
                quantum_registers[:-1],
            )
            diff_qubits = [0, 1, 2] if num_of_M == 1 else list(range(2 + num_of_M))
            qc.append(self.diffuser(len(diff_qubits)), diff_qubits)

        # --- Measurement ---
        qc.measure(pos_qubits, cbits[0:2])
        qc.measure(mov_qubits, cbits[2 : (2 + num_of_M)])

        # --- Simulation ---
        transpiled_qc = transpile(qc, self.simulator)
        result_counts = self.simulator.run(transpiled_qc).result().get_counts()

        # filtering out all the non-trivial solutions on the simulation output
        return [k for k in result_counts.keys() if k[-2:] == start_cell[cell_idx][len(start_cell[cell_idx]) :: -1]]

groverPlanner = GroverPlanner(obstacles_centroids.values())

# test
groverPlanner.find_path(start_cell=("00", "00"), target_cell=("00", "01"), m_type="macro")

# returns a list with the micro and macro displacement after the movement
def eval_displacement(old_centroid, new_centroid):
    old_c_micro, old_c_macro = old_centroid
    new_c_micro, new_c_macro = new_centroid
    return [manhattan(old_c_micro, new_c_micro), manhattan(old_c_macro, new_c_macro)]

# computes the movements required to get to the position encoded in new_alphas, which has the form [new_alphax, new_alphay]
#
# it also updates the new position of the robot, so there is no need to do that separately
#
def compute_path_to_position(robot, new_alphas):

    # starting centroid
    old_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)

    # assign new positions
    robot.alphax, robot.alphay = new_alphas
    robot.betax, robot.betay = c_ampl(robot.alphax), c_ampl(robot.alphay)

    # new centroid
    new_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)

    # displacement between the new and starting centroids
    displacement = eval_displacement(old_centroid, new_centroid)

    #
    # Planning the path according to the different possible displacement:
    #   - macro: the robot changes macro-cell but stays in the same micro-cell
    #   - micro: the robot changes micro-cell but stays in the same macro-cell
    #   - macro and micro: the robot changes both macro and micro cells
    #   - if the new position is in the same micro and macro cells nothing happens
    #

    # --- Macro movement ---
    if displacement[0] == 0 and displacement[1] != 0:
        print(f"{robot.name}: macro")
        evaluated_path = groverPlanner.find_path(old_centroid, new_centroid, m_type="macro")
        print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")

    # --- Micro movement ---
    elif displacement[0] != 0 and displacement[1] == 0:
        print(f"{robot.name}: micro")
        evaluated_path = groverPlanner.find_path(old_centroid, new_centroid, m_type="micro")
        print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")

    # --- Macro and micro movement ---
    elif displacement[0] != 0 and displacement[1] != 0:
        print(f"{robot.name}: macro-micro")
        # macro path
        evaluated_path = groverPlanner.find_path(old_centroid, new_centroid, m_type="macro")
        print(f"{robot.name} evaluated path (macro): {evaluated_path[0][:-2]}")

        # micro path
        evaluated_path = groverPlanner.find_path(old_centroid, new_centroid, m_type="micro")
        print(f"{robot.name} evaluated path (micro): {evaluated_path[0][:-2]}")

# check whether a given position ends up in a obstacle cell
def is_position_on_obstacle(new_alphas):
    new_pos_centroid = find_closest_centroid(centroids, c_ampl(new_alphas[0]), c_ampl(new_alphas[1]))
    return True if new_pos_centroid in obstacles_centroids.values() else False

def informed_alpha_reshuffling():

    # parse the coordinates of farthest robot from the target
    #
    # since this method is called when the robots are in a cluster (either at the start or ar the end
    # of a full procedure), it does not matter which robot is chosen, all being quite close to each other;
    # here, among all, I am choosing the farthest one with the idea of "maximum coverage" of the search space
    #
    farthest_robot = min(Robotx._registry, key=attrgetter('delta'))
    f_centroid = find_closest_centroid(centroids, farthest_robot.betax, farthest_robot.betay)
    f_coords = list(list(centroids.keys())[list(centroids.values()).index(f_centroid)])

    # handle the case where the robot lies in the same row/column of the target
    if f_coords[0] == T.x:
        f_coords[0] = f_coords[1]
    elif f_coords[1] == T.y:
        f_coords[1] = f_coords[0]

    # compute the new reshuffled position within the rectangle
    return [c_ampl(np.random.uniform(f_coords[0], T.x)), c_ampl(np.random.uniform(f_coords[1], T.y))]

result = all(i.delta < 0.8 for i in Robotx._registry)
print("Do all the robots have a reward lower than 0.8? : " + str(result))

# if True: reshuffle positions
# if False: do nothing

# reshuffling is considered invalid if the target position of the robot is the same as that of an obstacle;
# with this consideration, no computational time is wasted trying to find the path to an inaccessible cell

if result == True:
    for i in Robotx._registry:
        new_alphas = informed_alpha_reshuffling()

        # if new position ends up in an obstacle cell, recompute the alphas
        while is_position_on_obstacle(new_alphas):
            new_alphas = informed_alpha_reshuffling()

        compute_path_to_position(i, new_alphas)

# recalculate the rewards
def recalculate_rewards():
    for i in Robotx._registry:
        i.delta = reward(T, i.betax, i.betay)
        i.gamma = c_ampl(i.delta)
        print(f"{i.name} {i.delta:.2f}")

recalculate_rewards()
print(f"Average reward: {calculate_avg_reward()}")

plot_scatterplot("Reshuffled positions")

"""### 5.1 Circuit construction and execution"""

q = QuantumRegister(5, 'q') # qubits # changed to 9, formerly 15
m2 = ClassicalRegister(1, 'c1') # classical bits (separated is better)
m3 = ClassicalRegister(1, 'c2')
m4 = ClassicalRegister(1, 'c3')
meas = ClassicalRegister(3, 'meas')

qc3 = QuantumCircuit(q, m2, m3, m4) # to reach the target
qc4 = QuantumCircuit(q, m2, m3, m4) # to get back to the nest

# utility function to better visualize the statevectors we are working with
def print_formatted_vector(*args):
    for vector in args:
        print("[" + "".join(f"{val:.2f} " for val in vector).strip() + "]")

# in case of ties on delta score, the max() function outputs the first maximum item it processes
closest_robot = max(Robotx._registry, key=attrgetter('delta'))
print(f"Closest robot to the target: {closest_robot.name} {closest_robot.betax:.2f} {closest_robot.betay:.2f} {closest_robot.delta:.2f}")

# and then it enters the gate
vector0 = [closest_robot.alphax, closest_robot.betax]
vector1 = [closest_robot.alphay, closest_robot.betay]
vector3 = [closest_robot.gamma, closest_robot.delta]

normalized_v0 = vector0/np.linalg.norm(vector0)
normalized_v1 = vector1/np.linalg.norm(vector1)
normalized_v3 = vector3/np.linalg.norm(vector3)

print_formatted_vector(vector0, vector1, vector3)
print_formatted_vector(normalized_v0, normalized_v1, normalized_v3)

# direct initialization with amplitudes vector
qc3.initialize(normalized_v0, q[0])
qc3.initialize(normalized_v1, q[1])
qc3.initialize(normalized_v3, q[2])

!pip install pylatexenc

# this is the core code, and it is unchanged across time
#
qc3.barrier(q)

qc3.ccx(q[0],q[2],q[3])
qc3.ccx(q[1],q[2],q[4])

qc3.x(q[2])

qc3.ch(q[2],q[3])
qc3.ch(q[2],q[4])

qc3.x(q[2])

qc3.barrier(q)

# perform measurements and store them in classical bits
qc3.measure(q[2],m2[0])
qc3.measure(q[3],m3[0])
qc3.measure(q[4],m4[0])

# visualization of the ciruit
qc3.draw(output="mpl")

# --- Simulation ---
aer_sim = Aer.get_backend("aer_simulator")
transpiled_qc = transpile(qc3, aer_sim)
result = aer_sim.run(transpiled_qc).result()

counts = result.get_counts()
print("counts: ", counts)

plot_histogram(result.get_counts())

# calculate weighted outcome on the most probables outcomes of the circuit
def eval_outcome(most_prob_dict, n_outcome):
    mapped_weights = list(map(lambda res: int(res[n_outcome*2])*most_prob_dict[res], most_prob_dict))
    return sum(mapped_weights)/sum(most_prob_dict.values())

num_most_prob_states = 4

# extract the most probable states and then evaluate the outcome
most_prob_dict = dict(nlargest(num_most_prob_states, counts.items(), key=itemgetter(1)))
print(f"{num_most_prob_states} most probable states: {most_prob_dict}")

outcome0, outcome1 = eval_outcome(most_prob_dict, 0), eval_outcome(most_prob_dict, 1)

print(f"outcome0: {outcome0:.2f}\noutcome1: {outcome1:.2f}")

for i in Robotx._registry:
    print(f"{i.name} {i.betax:.2f} {i.betay:.2f}")
    if (i.delta != closest_robot.delta or all(i.delta == j.delta for j in Robotx._registry)): # excluding the robot that entered the gate
        compute_path_to_position(i, [c_ampl(outcome1), c_ampl(outcome0)]) # update on alphas is done within the function

plot_scatterplot("Closest robot position update")

# former rewards
for i in  Robotx._registry:
    print(f"before the gate: {i.name} {i.delta:.2f}")

# new rewards
recalculate_rewards()
print(f"Average reward: {calculate_avg_reward()}")

# parse the most successful robot
max_attr_ = max(Robotx._registry, key=attrgetter('delta'))
print(f"max_attr_.delta: {max_attr_.delta:.2f}")

for i in Robotx._registry:
    if (i.delta == max_attr_.delta):
        print(f"Most successful robot: {i.name} {i.betax:.2f} {i.betay:.2f} {i.delta:.2f}") # and then it enters the gate

for j in Robotx._registry:
    # to get other robots following it:
    if (j != max_attr_):
        compute_path_to_position(j, [max_attr_.alphax + np.random.uniform(0,0.01), max_attr_.alphay + np.random.uniform(0,0.01)])

recalculate_rewards()
print(f"Average reward: {calculate_avg_reward()}")

plot_scatterplot("Final configuration")

for i in Robotx._registry:
    if (i.delta < 0.8):

        new_alphas = informed_alpha_reshuffling()

        # if new position ends up in an obstacle cell, recompute the alphas
        while is_position_on_obstacle(new_alphas):
                new_alphas = informed_alpha_reshuffling()

        compute_path_to_position(i, new_alphas)

recalculate_rewards()

plot_scatterplot("Second reshuffle")

# parse the most successful robot
max_attr_ = max(Robotx._registry, key=attrgetter('delta'))
print(f"max_attr_.delta: {max_attr_.delta:.2f}")

for i in Robotx._registry:
    if (i.delta == max_attr_.delta):
        print(f"Most successful robot: {i.name} {i.betax:.2f} {i.betay:.2f} {i.delta:.2f}") # and then it enters the gate

for j in Robotx._registry:
    # to get other robots following it:
    if (j != max_attr_):
        compute_path_to_position(j, [max_attr_.alphax + np.random.uniform(0,0.01), max_attr_.alphay + np.random.uniform(0,0.01)])

recalculate_rewards()

plot_scatterplot("Intermediate step")

"""Here, the quantum swarm algorithm is invoked yet another time."""

# in case of ties on delta score, the max() function outputs the first maximum item it processes
closest_robot = max(Robotx._registry, key=attrgetter('delta'))
print(f"Closest robot to the target: {closest_robot.name} {closest_robot.betax:.2f} {closest_robot.betay:.2f} {closest_robot.delta:.2f}")

# and then it enters the gate
vector0 = [closest_robot.alphax, closest_robot.betax]
vector1 = [closest_robot.alphay, closest_robot.betay]
vector3 = [closest_robot.gamma, closest_robot.delta]

normalized_v0 = vector0/np.linalg.norm(vector0)
normalized_v1 = vector1/np.linalg.norm(vector1)
normalized_v3 = vector3/np.linalg.norm(vector3)

print_formatted_vector(vector0, vector1, vector3)
print_formatted_vector(normalized_v0, normalized_v1, normalized_v3)

# direct initialization with amplitudes vector
qc3.initialize(normalized_v0, q[0])
qc3.initialize(normalized_v1, q[1])
qc3.initialize(normalized_v3, q[2])

# this is the core code, and it is unchanged across time

qc3.barrier(q)

qc3.ccx(q[0],q[2],q[3])
qc3.ccx(q[1],q[2],q[4])

qc3.x(q[2])

qc3.ch(q[2],q[3])
qc3.ch(q[2],q[4])

qc3.x(q[2])

qc3.barrier(q)

# perform measurements and store them in classical bits

qc3.measure(q[2],m2[0])
qc3.measure(q[3],m3[0])
qc3.measure(q[4],m4[0])

# --- Simulation ---
transpiled_qc = transpile(qc3, aer_sim)
result = aer_sim.run(transpiled_qc).result()

counts = result.get_counts()
print("counts: ", counts)

plot_histogram(result.get_counts())

# extract the most probable states and then evaluate the outcome
most_prob_dict = dict(nlargest(num_most_prob_states, counts.items(), key=itemgetter(1)))
print(f"{num_most_prob_states} most probable states: {most_prob_dict}")

outcome0, outcome1 = eval_outcome(most_prob_dict, 0), eval_outcome(most_prob_dict, 1)

print(f"outcome0: {outcome0:.2f}\noutcome1: {outcome1:.2f}")

for i in Robotx._registry:
    print(f"{i.name} {i.betax:.2f} {i.betay:.2f}")
    if (i.delta != closest_robot.delta or all(i.delta == j.delta for j in Robotx._registry)): # excluding the robot that entered the gate
        compute_path_to_position(i, [c_ampl(outcome1), c_ampl(outcome0)]) # update on alphas is done within the function

# former rewards
for i in  Robotx._registry:
    print(f"before the gate: {i.name} {i.delta:.2f}")

# new rewards
print("Rewards after the gate: ")
recalculate_rewards()

# parse the most successful robot
max_attr_ = max(Robotx._registry, key=attrgetter('delta'))
print(f"max_attr_.delta: {max_attr_.delta:.2f}")

for i in Robotx._registry:
    if (i.delta == max_attr_.delta):
        print(f"Most successful robot: {i.name} {i.betax:.2f} {i.betay:.2f} {i.delta:.2f}") # and then it enters the gate

for j in Robotx._registry:
    # to get other robots following it:
    if (j != max_attr_):
        compute_path_to_position(j, [max_attr_.alphax + np.random.uniform(0,0.01), max_attr_.alphay + np.random.uniform(0,0.01)])

recalculate_rewards()

plot_scatterplot("Final after second gate")

print("\n=== SIMULATION COMPLETE ===")
print(f"Total obstacles in environment: {len(Obstacle._registry)}")
print("Obstacle positions:")
for obs in Obstacle._registry:
    print(f"  {obs.name}: ({obs.x:.3f}, {obs.y:.3f})")
print(f"\nFinal average reward: {calculate_avg_reward():.4f}")
print(f"Target location: ({T.x:.3f}, {T.y:.3f})")
```

```python
# -*- coding: utf-8 -*-
"""
Quantum Swarm Robotics - Experimental Framework
Refactored code to enable parametric experiments
"""

import numpy as np
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from operator import attrgetter, itemgetter
from scipy.spatial.distance import cityblock
from heapq import nlargest
from typing import List, Tuple, Dict, Optional


class Target:
    """Represents the target location"""
    def __init__(self, name: str, x: float, y: float):
        self.name = name
        self.x = x
        self.y = y


class Obstacle:
    """Represents an obstacle in the environment"""
    def __init__(self, name: str, x: float, y: float):
        self.name = name
        self.x = x
        self.y = y


class Robot:
    """Represents a robot in the swarm"""
    def __init__(self, name: str, alphax: float, alphay: float, target: Target):
        self.name = name
        self.alphax = alphax
        self.betax = self._c_ampl(alphax)
        self.alphay = alphay
        self.betay = self._c_ampl(alphay)
        self.delta = self._reward(target, self.betax, self.betay)
        self.gamma = self._c_ampl(self.delta)
    
    @staticmethod
    def _c_ampl(ampl: float) -> float:
        """Compute complementary amplitude"""
        return np.sqrt(1 - ampl**2)
    
    @staticmethod
    def _reward(target: Target, betax: float, betay: float) -> float:
        """Calculate reward based on distance to target"""
        return 1 - (((target.x - betax)**2 + (target.y - betay)**2)**0.5) / np.sqrt(2)
    
    def update_position(self, new_alphax: float, new_alphay: float, target: Target):
        """Update robot position and recalculate reward"""
        self.alphax = new_alphax
        self.alphay = new_alphay
        self.betax = self._c_ampl(new_alphax)
        self.betay = self._c_ampl(new_alphay)
        self.delta = self._reward(target, self.betax, self.betay)
        self.gamma = self._c_ampl(self.delta)


class GroverPlanner:
    """Quantum path planner using Grover's algorithm"""
    def __init__(self, obstacles: List[Tuple]):
        self.simulator = Aer.get_backend("aer_simulator")
        self.obstacles = obstacles
    
    def XOR(self, qc, a, b, output):
        qc.cx(a, output)
        qc.cx(b, output)
    
    def encode_cell(self, qc, cell, qubits):
        for idx, bit in enumerate(cell):
            if int(bit):
                qc.x(qubits[idx])
    
    def encode_target(self, qc, target_cell, qubits):
        for idx, bit in enumerate(target_cell):
            if not int(bit):
                qc.x(qubits[idx])
    
    def obstacles_in_map(self, target_cell):
        return any(target_cell[1] == macro_cell for macro_cell in list(map(lambda ob_t: ob_t[1], self.obstacles)))
    
    def diffuser(self, nqubits):
        diff = QuantumCircuit(nqubits)
        for qubit in range(nqubits):
            diff.h(qubit)
        for qubit in range(nqubits):
            diff.x(qubit)
        diff.h(nqubits - 1)
        diff.mcx(list(range(nqubits - 1)), nqubits - 1, mode='noancilla')
        diff.h(nqubits - 1)
        for qubit in range(nqubits):
            diff.x(qubit)
        for qubit in range(nqubits):
            diff.h(qubit)
        U_s = diff.to_gate()
        U_s.name = "U$_\\psi$"
        return U_s
    
    def half_oracle(self, num_of_M, obstacles_in_map, quantum_registers):
        if num_of_M == 1:
            ho = QuantumCircuit(*quantum_registers)
            pos_qubits, mov_qubits, c, output = quantum_registers
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])
        elif num_of_M == 2:
            ho = QuantumCircuit(*quantum_registers)
            if obstacles_in_map:
                pos_qubits, mov_qubits, c, o, caux, out, output = quantum_registers
            else:
                pos_qubits, mov_qubits, c, output = quantum_registers
            
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])
            
            if obstacles_in_map:
                for obstacle in self.obstacles:
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])
                    ho.ccx(c[0], c[1], o)
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])
            
            self.XOR(ho, c[1], mov_qubits[1], c[2])
            ho.x(c[2])
            self.XOR(ho, c[0], mov_qubits[1], c[3])
            
            if obstacles_in_map:
                ho.ccx(c[2], o, caux[0])
                ho.x(caux[0])
                ho.ccx(c[3], o, caux[1])
                ho.x(caux[1])
                ho.x(o)
                ho.ccx(pos_qubits[0], o, caux[2])
                ho.x(caux[2])
                ho.ccx(pos_qubits[1], o, caux[3])
                ho.x(caux[3])
                ho.ccx(caux[0], caux[2], out[0])
                ho.x(out[0])
                ho.ccx(caux[1], caux[3], out[1])
                ho.x(out[1])
        
        HalfOracle = ho.to_gate()
        HalfOracle.name = "HO"
        return HalfOracle
    
    def oracle(self, num_of_M, target_cell, m_type, obstacles_in_map, qc, quantum_registers):
        halforacle = self.half_oracle(num_of_M, obstacles_in_map, quantum_registers)
        additional_regs = []
        
        if obstacles_in_map:
            pos_qubits, mov_qubits, c, *additional_regs, output = quantum_registers
        else:
            pos_qubits, mov_qubits, c, output = quantum_registers
        
        qc.append(
            halforacle,
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + ((additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:]) if additional_regs else [])
            + output[:],
        )
        
        test_qubits = c[-2:] if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map) else additional_regs[2]
        tcell_idx = 0 if m_type == "micro" else 1
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])
        qc.ccx(test_qubits[0], test_qubits[1], output)
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])
        
        qc.append(
            halforacle.inverse(),
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + ((additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:]) if additional_regs else [])
            + output[:],
        )
    
    def find_path(self, start_cell, target_cell, m_type):
        cell_idx = 0 if m_type == "micro" else 1
        num_of_M = manhattan(start_cell[cell_idx], target_cell[cell_idx])
        
        pos_qubits = QuantumRegister(2, name="r")
        mov_qubits = QuantumRegister(num_of_M, name="m")
        c = QuantumRegister(2 * num_of_M, name="c")
        o = QuantumRegister(1, name="o")
        caux = QuantumRegister(4, name="caux")
        out = QuantumRegister(2, name="out")
        output = QuantumRegister(1, name="q")
        cbits = ClassicalRegister(2 + num_of_M, name="cbits")
        
        obstacles_in_map = self.obstacles_in_map(target_cell) and m_type == "micro"
        
        base_registers = [pos_qubits, mov_qubits, c, output, cbits]
        additional_registers = base_registers.copy()
        additional_registers[3:3] = [o, caux, out]
        quantum_registers = (
            base_registers if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map)
            else additional_registers
        )
        
        qc = QuantumCircuit(*quantum_registers)
        qc.h(pos_qubits)
        qc.h(mov_qubits)
        qc.initialize([1, -1] / np.sqrt(2), output)
        
        num_iterations = 1
        for _ in range(num_iterations):
            self.oracle(num_of_M, target_cell, m_type, obstacles_in_map, qc, quantum_registers[:-1])
            diff_qubits = [0, 1, 2] if num_of_M == 1 else list(range(2 + num_of_M))
            qc.append(self.diffuser(len(diff_qubits)), diff_qubits)
        
        qc.measure(pos_qubits, cbits[0:2])
        qc.measure(mov_qubits, cbits[2:(2 + num_of_M)])
        
        transpiled_qc = transpile(qc, self.simulator)
        result_counts = self.simulator.run(transpiled_qc).result().get_counts()
        
        return [k for k in result_counts.keys() if k[-2:] == start_cell[cell_idx][len(start_cell[cell_idx]):: -1]]


def find_centroids(dim_map: int) -> Dict:
    """Generate centroids for the map grid"""
    unit_displacement = 1 / (dim_map * 2)
    centroids_coords = []
    for i in range(dim_map, 0, -1):
        row = []
        for j in range(dim_map):
            row.append((j * unit_displacement * 2 + unit_displacement, 
                       i * unit_displacement * 2 - unit_displacement))
        centroids_coords.append(row)
    
    centroids_dict = {
        t: (f"{str(i%2)}{str(j%2)}", f"{0 if t[1] > 0.5 else 1}{0 if t[0] < 0.5 else 1}")
        for i, row in enumerate(centroids_coords) for j, t in enumerate(row)
    }
    return centroids_dict


def find_closest_centroid(centroids: Dict, x: float, y: float) -> Tuple:
    """Find the closest centroid to a given position"""
    centroids_coords = np.array(list(centroids.keys()))
    distances = np.linalg.norm(centroids_coords - np.array([x, y]), axis=1)
    return centroids[tuple(centroids_coords[np.argmin(distances)])]


def manhattan(start_cell: str, target_cell: str) -> int:
    """Calculate Manhattan distance between cells"""
    start_cell_v = [int(b) for b in start_cell]
    target_cell_v = [int(b) for b in target_cell]
    return int(cityblock(start_cell_v, target_cell_v).item())


def eval_displacement(old_centroid: Tuple, new_centroid: Tuple) -> List[int]:
    """Calculate displacement between centroids"""
    old_c_micro, old_c_macro = old_centroid
    new_c_micro, new_c_macro = new_centroid
    return [manhattan(old_c_micro, new_c_micro), manhattan(old_c_macro, new_c_macro)]


def compute_path_to_position(robot: Robot, new_alphas: List[float], 
                            centroids: Dict, grover_planner: GroverPlanner,
                            target: Target, verbose: bool = False):
    """Compute and execute path for robot to new position"""
    old_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)
    robot.update_position(new_alphas[0], new_alphas[1], target)
    new_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)
    displacement = eval_displacement(old_centroid, new_centroid)
    
    if displacement[0] == 0 and displacement[1] != 0:
        if verbose:
            print(f"{robot.name}: macro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="macro")
        if verbose:
            print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")
    elif displacement[0] != 0 and displacement[1] == 0:
        if verbose:
            print(f"{robot.name}: micro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="micro")
        if verbose:
            print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")
    elif displacement[0] != 0 and displacement[1] != 0:
        if verbose:
            print(f"{robot.name}: macro-micro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="macro")
        if verbose:
            print(f"{robot.name} evaluated path (macro): {evaluated_path[0][:-2]}")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="micro")
        if verbose:
            print(f"{robot.name} evaluated path (micro): {evaluated_path[0][:-2]}")


def is_position_on_obstacle(new_alphas: List[float], centroids: Dict, 
                           obstacles_centroids: Dict) -> bool:
    """Check if position collides with obstacle"""
    c_ampl = lambda a: np.sqrt(1 - a**2)
    new_pos_centroid = find_closest_centroid(centroids, c_ampl(new_alphas[0]), c_ampl(new_alphas[1]))
    return new_pos_centroid in obstacles_centroids.values()


def informed_alpha_reshuffling(robots: List[Robot], target: Target) -> List[float]:
    """Generate new position within search rectangle"""
    c_ampl = lambda a: np.sqrt(1 - a**2)
    centroids = find_centroids(4)
    
    farthest_robot = min(robots, key=attrgetter('delta'))
    f_centroid = find_closest_centroid(centroids, farthest_robot.betax, farthest_robot.betay)
    f_coords = list(list(centroids.keys())[list(centroids.values()).index(f_centroid)])
    
    if f_coords[0] == target.x:
        f_coords[0] = f_coords[1]
    elif f_coords[1] == target.y:
        f_coords[1] = f_coords[0]
    
    return [c_ampl(np.random.uniform(f_coords[0], target.x)), 
            c_ampl(np.random.uniform(f_coords[1], target.y))]


def calculate_avg_reward(robots: List[Robot]) -> float:
    """Calculate average reward across all robots"""
    return np.mean([robot.delta for robot in robots])


def run_quantum_swarm_experiment(
    num_robots: int = 10,
    obstacle_positions: List[Tuple[float, float]] = None,
    target_position: Tuple[float, float] = (0.875, 0.125),
    delta_threshold: float = 0.8,
    starting_position: Tuple[float, float] = (0.125, 0.875),
    starting_radius: float = 0.05,
    num_reshuffles: int = 2,
    num_gate_applications: int = 2,
    verbose: bool = False,
    plot: bool = False
) -> Dict:
    """
    Run a complete quantum swarm robotics experiment.
    
    Parameters:
    -----------
    num_robots : int
        Number of robots in the swarm
    obstacle_positions : List[Tuple[float, float]]
        List of (x, y) coordinates for obstacles
    target_position : Tuple[float, float]
        (x, y) coordinates of the target
    delta_threshold : float
        Reward threshold for triggering reshuffle
    starting_position : Tuple[float, float]
        Center of initial robot cluster
    starting_radius : float
        Radius of initial robot distribution
    num_reshuffles : int
        Number of reshuffle operations to perform
    num_gate_applications : int
        Number of times to apply the quantum gate
    verbose : bool
        Print detailed progress information
    plot : bool
        Generate visualization plots
    
    Returns:
    --------
    Dict containing experimental results and reward history
    """
    
    # Setup
    if obstacle_positions is None:
        obstacle_positions = [(0.125, 0.125), (0.875, 0.375), (0.625, 0.125)]
    
    target = Target("T", target_position[0], target_position[1])
    obstacles = [Obstacle(f"O{i+1}", pos[0], pos[1]) for i, pos in enumerate(obstacle_positions)]
    
    # Initialize robots
    c_ampl = lambda a: np.sqrt(1 - a**2)
    a_x = c_ampl(starting_position[0] - starting_radius)
    a_y = c_ampl(starting_position[0] + starting_radius)
    b_x = c_ampl(starting_position[1] - starting_radius)
    b_y = c_ampl(starting_position[1] + starting_radius)
    
    robots = []
    for i in range(num_robots):
        robot = Robot(f'R{i}', 
                     np.random.uniform(a_x, a_y), 
                     np.random.uniform(b_x, b_y),
                     target)
        robots.append(robot)
    
    # Setup environment
    centroids = find_centroids(4)
    obstacles_centroids = {
        coord: v for (coord, v) in centroids.items() 
        for o in obstacles if coord == (o.x, o.y)
    }
    grover_planner = GroverPlanner(list(obstacles_centroids.values()))
    
    # Track rewards at different stages
    rewards_history = {
        't0_initial': calculate_avg_reward(robots),
        'reshuffles': [],
        'gates': [],
        'final': None
    }
    
    if verbose:
        print(f"Initial average reward: {rewards_history['t0_initial']:.4f}")
    
    # Main experiment loop
    for reshuffle_idx in range(num_reshuffles):
        # Check if reshuffle needed
        if all(r.delta < delta_threshold for r in robots):
            if verbose:
                print(f"\n=== Reshuffle {reshuffle_idx + 1} ===")
            
            for robot in robots:
                new_alphas = informed_alpha_reshuffling(robots, target)
                while is_position_on_obstacle(new_alphas, centroids, obstacles_centroids):
                    new_alphas = informed_alpha_reshuffling(robots, target)
                compute_path_to_position(robot, new_alphas, centroids, grover_planner, target, verbose)
            
            rewards_history['reshuffles'].append(calculate_avg_reward(robots))
            if verbose:
                print(f"Average reward after reshuffle: {rewards_history['reshuffles'][-1]:.4f}")
        
        # Apply quantum gate
        if reshuffle_idx < num_gate_applications:
            if verbose:
                print(f"\n=== Quantum Gate {reshuffle_idx + 1} ===")
            
            closest_robot = max(robots, key=attrgetter('delta'))
            
            # Quantum circuit setup
            q = QuantumRegister(5, 'q')
            m2 = ClassicalRegister(1, 'c1')
            m3 = ClassicalRegister(1, 'c2')
            m4 = ClassicalRegister(1, 'c3')
            qc = QuantumCircuit(q, m2, m3, m4)
            
            # Initialize quantum state
            vector0 = np.array([closest_robot.alphax, closest_robot.betax])
            vector1 = np.array([closest_robot.alphay, closest_robot.betay])
            vector3 = np.array([closest_robot.gamma, closest_robot.delta])
            
            qc.initialize(vector0 / np.linalg.norm(vector0), q[0])
            qc.initialize(vector1 / np.linalg.norm(vector1), q[1])
            qc.initialize(vector3 / np.linalg.norm(vector3), q[2])
            
            # Apply quantum operations
            qc.barrier(q)
            qc.ccx(q[0], q[2], q[3])
            qc.ccx(q[1], q[2], q[4])
            qc.x(q[2])
            qc.ch(q[2], q[3])
            qc.ch(q[2], q[4])
            qc.x(q[2])
            qc.barrier(q)
            
            # Measure
            qc.measure(q[2], m2[0])
            qc.measure(q[3], m3[0])
            qc.measure(q[4], m4[0])
            
            # Simulate
            aer_sim = Aer.get_backend("aer_simulator")
            transpiled_qc = transpile(qc, aer_sim)
            result = aer_sim.run(transpiled_qc).result()
            counts = result.get_counts()
            
            # Extract outcomes
            num_most_prob_states = min(4, len(counts))
            most_prob_dict = dict(nlargest(num_most_prob_states, counts.items(), key=itemgetter(1)))
            
            def eval_outcome(most_prob_dict, n_outcome):
                mapped_weights = [int(res[n_outcome * 2]) * most_prob_dict[res] for res in most_prob_dict]
                return sum(mapped_weights) / sum(most_prob_dict.values())
            
            outcome0, outcome1 = eval_outcome(most_prob_dict, 0), eval_outcome(most_prob_dict, 1)
            
            # Update other robots
            for robot in robots:
                if robot.delta != closest_robot.delta or all(r.delta == robots[0].delta for r in robots):
                    compute_path_to_position(robot, [c_ampl(outcome1), c_ampl(outcome0)], 
                                           centroids, grover_planner, target, verbose)
            
            # Cluster around best robot
            max_robot = max(robots, key=attrgetter('delta'))
            for robot in robots:
                if robot != max_robot:
                    compute_path_to_position(
                        robot,
                        [max_robot.alphax + np.random.uniform(0, 0.01),
                         max_robot.alphay + np.random.uniform(0, 0.01)],
                        centroids, grover_planner, target, verbose
                    )
            
            rewards_history['gates'].append(calculate_avg_reward(robots))
            if verbose:
                print(f"Average reward after gate: {rewards_history['gates'][-1]:.4f}")
    
    # Final reward
    rewards_history['final'] = calculate_avg_reward(robots)
    
    if verbose:
        print(f"\n=== Experiment Complete ===")
        print(f"Final average reward: {rewards_history['final']:.4f}")
        print(f"Total obstacles: {len(obstacles)}")
    
    # Plot if requested
    if plot:
        plot_experiment_results(robots, target, obstacles)
    
    return {
        'robots': robots,
        'target': target,
        'obstacles': obstacles,
        'rewards_history': rewards_history,
        'num_robots': num_robots,
        'delta_threshold': delta_threshold
    }


def plot_experiment_results(robots: List[Robot], target: Target, 
                           obstacles: List[Obstacle], title: str = "Experiment Results"):
    """Visualize the final state of the experiment"""
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xticks(np.arange(0, 1, 0.25))
    ax.set_yticks(np.arange(0, 1, 0.25))
    plt.grid()
    
    for i, robot in enumerate(robots):
        robot_color = f"#{30 + i * 20:02x}{200 + i * 5:02x}15"
        plt.scatter(robot.betax, robot.betay, s=300, marker='o', color=robot_color, alpha=0.7)
    
    plt.scatter(target.x, target.y, s=400, marker='*', color='turquoise', label='Target')
    
    for obstacle in obstacles:
        plt.scatter(obstacle.x, obstacle.y, s=400, marker='p', color='red', alpha=0.7)
    
    plt.axis([0, 1, 0, 1])
    plt.title(title)
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.legend()
    plt.show()


def get_mean_rewards_at_stages(rewards_history: Dict) -> Dict[str, float]:
    """
    Extract mean rewards at different stages of the experiment.
    
    Parameters:
    -----------
    rewards_history : Dict
        The rewards_history from run_quantum_swarm_experiment
    
    Returns:
    --------
    Dict with keys 't0', 't1', 't2', 't3' representing:
        t0: Initial state
        t1: After first reshuffle
        t2: After first logic gate
        t3: Final state
    """
    result = {
        't0': rewards_history['t0_initial'],
        't1': rewards_history['reshuffles'][0] if rewards_history['reshuffles'] else None,
        't2': rewards_history['gates'][0] if rewards_history['gates'] else None,
        't3': rewards_history['final']
    }
    
    return result


def run_multiple_experiments(
    num_experiments: int = 10,
    **experiment_params
) -> List[Dict]:
    """
    Run multiple experiments and collect results.
    
    Parameters:
    -----------
    num_experiments : int
        Number of experiments to run
    **experiment_params : 
        Parameters to pass to run_quantum_swarm_experiment
    
    Returns:
    --------
    List of experiment results
    """
    results = []
    
    for i in range(num_experiments):
        print(f"\n{'='*50}")
        print(f"Running Experiment {i + 1}/{num_experiments}")
        print(f"{'='*50}")
        
        result = run_quantum_swarm_experiment(
            verbose=experiment_params.get('verbose', False),
            plot=experiment_params.get('plot', False),
            **{k: v for k, v in experiment_params.items() if k not in ['verbose', 'plot']}
        )
        
        results.append(result)
        
        # Print stage rewards
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        print(f"\nStage Rewards:")
        for stage, reward in stage_rewards.items():
            print(f"  {stage}: {reward:.4f}" if reward is not None else f"  {stage}: N/A")
    
    return results


def aggregate_experiment_statistics(results: List[Dict]) -> Dict:
    """
    Aggregate statistics across multiple experiments.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results from run_multiple_experiments
    
    Returns:
    --------
    Dictionary containing aggregated statistics
    """
    all_stage_rewards = {
        't0': [],
        't1': [],
        't2': [],
        't3': []
    }
    
    for result in results:
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        for stage, reward in stage_rewards.items():
            if reward is not None:
                all_stage_rewards[stage].append(reward)
    
    statistics = {}
    for stage, rewards in all_stage_rewards.items():
        if rewards:
            statistics[stage] = {
                'mean': np.mean(rewards),
                'std': np.std(rewards),
                'min': np.min(rewards),
                'max': np.max(rewards),
                'median': np.median(rewards)
            }
        else:
            statistics[stage] = None
    
    return statistics


def plot_reward_progression(results: List[Dict], title: str = "Reward Progression Across Experiments"):
    """
    Plot reward progression across multiple experiments.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results
    title : str
        Plot title
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Individual experiment trajectories
    for idx, result in enumerate(results):
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        stages = ['t0', 't1', 't2', 't3']
        rewards = [stage_rewards.get(s) for s in stages]
        
        # Filter out None values
        valid_stages = [s for s, r in zip(stages, rewards) if r is not None]
        valid_rewards = [r for r in rewards if r is not None]
        
        ax1.plot(valid_stages, valid_rewards, marker='o', alpha=0.5, label=f'Exp {idx+1}')
    
    ax1.set_xlabel('Stage')
    ax1.set_ylabel('Average Reward')
    ax1.set_title('Individual Experiment Trajectories')
    ax1.grid(True, alpha=0.3)
    ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # Aggregated statistics
    stats = aggregate_experiment_statistics(results)
    stages = ['t0', 't1', 't2', 't3']
    means = [stats[s]['mean'] if stats[s] else None for s in stages]
    stds = [stats[s]['std'] if stats[s] else 0 for s in stages]
    
    valid_stages = [s for s, m in zip(stages, means) if m is not None]
    valid_means = [m for m in means if m is not None]
    valid_stds = [std for m, std in zip(means, stds) if m is not None]
    
    ax2.errorbar(valid_stages, valid_means, yerr=valid_stds, 
                 marker='o', markersize=10, linewidth=2, capsize=5,
                 label='Mean ± Std')
    ax2.fill_between(range(len(valid_means)), 
                      np.array(valid_means) - np.array(valid_stds),
                      np.array(valid_means) + np.array(valid_stds),
                      alpha=0.3)
    
    ax2.set_xlabel('Stage')
    ax2.set_ylabel('Average Reward')
    ax2.set_title('Aggregated Statistics (Mean ± Std)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()


def compare_experiment_configurations(configurations: List[Dict]) -> None:
    """
    Compare results across different experiment configurations.
    
    Parameters:
    -----------
    configurations : List[Dict]
        List of configuration dictionaries, each containing:
        - 'name': configuration name
        - 'params': parameters for run_multiple_experiments
        - 'num_experiments': number of experiments to run
    
    Example:
    --------
    configs = [
        {
            'name': 'Few Obstacles',
            'params': {'obstacle_positions': [(0.125, 0.125)]},
            'num_experiments': 5
        },
        {
            'name': 'Many Obstacles',
            'params': {'obstacle_positions': [(0.125, 0.125), (0.625, 0.625), (0.875, 0.375)]},
            'num_experiments': 5
        }
    ]
    compare_experiment_configurations(configs)
    """
    all_results = {}
    
    for config in configurations:
        print(f"\n{'='*60}")
        print(f"Testing Configuration: {config['name']}")
        print(f"{'='*60}")
        
        results = run_multiple_experiments(
            num_experiments=config['num_experiments'],
            **config['params']
        )
        
        all_results[config['name']] = results
        
        # Print summary statistics
        stats = aggregate_experiment_statistics(results)
        print(f"\n{config['name']} - Summary Statistics:")
        for stage, stat in stats.items():
            if stat:
                print(f"  {stage}: {stat['mean']:.4f} ± {stat['std']:.4f} "
                      f"(min: {stat['min']:.4f}, max: {stat['max']:.4f})")
    
    # Comparative visualization
    fig, axes = plt.subplots(1, len(configurations), figsize=(6*len(configurations), 5))
    if len(configurations) == 1:
        axes = [axes]
    
    for idx, (config_name, results) in enumerate(all_results.items()):
        stats = aggregate_experiment_statistics(results)
        stages = ['t0', 't1', 't2', 't3']
        means = [stats[s]['mean'] if stats[s] else None for s in stages]
        stds = [stats[s]['std'] if stats[s] else 0 for s in stages]
        
        valid_stages = [s for s, m in zip(stages, means) if m is not None]
        valid_means = [m for m in means if m is not None]
        valid_stds = [std for m, std in zip(means, stds) if m is not None]
        
        axes[idx].errorbar(valid_stages, valid_means, yerr=valid_stds,
                          marker='o', markersize=10, linewidth=2, capsize=5)
        axes[idx].fill_between(range(len(valid_means)),
                               np.array(valid_means) - np.array(valid_stds),
                               np.array(valid_means) + np.array(valid_stds),
                               alpha=0.3)
        axes[idx].set_xlabel('Stage')
        axes[idx].set_ylabel('Average Reward')
        axes[idx].set_title(config_name)
        axes[idx].grid(True, alpha=0.3)
        axes[idx].set_ylim([0, 1])
    
    plt.tight_layout()
    plt.show()
    
    return all_results


def export_results_to_csv(results: List[Dict], filename: str = "experiment_results.csv"):
    """
    Export experiment results to CSV file.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results
    filename : str
        Output CSV filename
    """
    import csv
    
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['experiment_id', 'num_robots', 'num_obstacles', 'delta_threshold',
                     't0_reward', 't1_reward', 't2_reward', 't3_reward',
                     'improvement_t0_to_t3']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for idx, result in enumerate(results):
            stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
            
            row = {
                'experiment_id': idx + 1,
                'num_robots': result['num_robots'],
                'num_obstacles': len(result['obstacles']),
                'delta_threshold': result['delta_threshold'],
                't0_reward': stage_rewards.get('t0'),
                't1_reward': stage_rewards.get('t1'),
                't2_reward': stage_rewards.get('t2'),
                't3_reward': stage_rewards.get('t3'),
                'improvement_t0_to_t3': (stage_rewards.get('t3', 0) - stage_rewards.get('t0', 0))
            }
            writer.writerow(row)
    
    print(f"Results exported to {filename}")


# Example usage demonstrations:
if __name__ == "__main__":
    
    print("="*60)
    print("EXAMPLE 1: Single Experiment with Custom Parameters")
    print("="*60)
    
    result = run_quantum_swarm_experiment(
        num_robots=10,
        obstacle_positions=[(0.125, 0.125), (0.875, 0.375), (0.625, 0.125)],
        target_position=(0.875, 0.125),
        delta_threshold=0.8,
        starting_position=(0.125, 0.875),
        starting_radius=0.05,
        num_reshuffles=2,
        num_gate_applications=2,
        verbose=True,
        plot=True
    )
    
    stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
    print("\nMean Rewards at Different Stages:")
    for stage, reward in stage_rewards.items():
        print(f"  {stage}: {reward:.4f}" if reward else f"  {stage}: N/A")
    
    
    print("\n" + "="*60)
    print("EXAMPLE 2: Multiple Experiments with Same Configuration")
    print("="*60)
    
    experiments = run_multiple_experiments(
        num_experiments=3,
        num_robots=8,
        delta_threshold=0.75,
        obstacle_positions=[(0.125, 0.125), (0.625, 0.625)],
        verbose=False,
        plot=False
    )
    
    # Show aggregated statistics
    stats = aggregate_experiment_statistics(experiments)
    print("\nAggregated Statistics Across Experiments:")
    for stage, stat in stats.items():
        if stat:
            print(f"\n{stage}:")
            print(f"  Mean: {stat['mean']:.4f}")
            print(f"  Std:  {stat['std']:.4f}")
            print(f"  Min:  {stat['min']:.4f}")
            print(f"  Max:  {stat['max']:.4f}")
    
    # Plot progression
    plot_reward_progression(experiments, "Reward Progression - 8 Robots, 2 Obstacles")
    
    
    print("\n" + "="*60)
    print("EXAMPLE 3: Comparing Different Configurations")
    print("="*60)
    
    configurations = [
        {
            'name': '5 Robots - Low Threshold',
            'params': {
                'num_robots': 5,
                'delta_threshold': 0.6,
                'obstacle_positions': [(0.125, 0.125), (0.625, 0.625)]
            },
            'num_experiments': 3
        },
        {
            'name': '15 Robots - High Threshold',
            'params': {
                'num_robots': 15,
                'delta_threshold': 0.9,
                'obstacle_positions': [(0.125, 0.125), (0.625, 0.625)]
            },
            'num_experiments': 3
        }
    ]
    
    comparison_results = compare_experiment_configurations(configurations)
    
    
    print("\n" + "="*60)
    print("EXAMPLE 4: Export Results to CSV")
    print("="*60)
    
    export_results_to_csv(experiments, "quantum_swarm_results.csv")
```

```python
# -*- coding: utf-8 -*-
"""
Quantum Swarm Robotics - Experimental Framework
Refactored code to enable parametric experiments
"""

import numpy as np
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from operator import attrgetter, itemgetter
from scipy.spatial.distance import cityblock
from heapq import nlargest
from typing import List, Tuple, Dict, Optional


class Target:
    """Represents the target location"""
    def __init__(self, name: str, x: float, y: float):
        self.name = name
        self.x = x
        self.y = y


class Obstacle:
    """Represents an obstacle in the environment"""
    def __init__(self, name: str, x: float, y: float):
        self.name = name
        self.x = x
        self.y = y


class Robot:
    """Represents a robot in the swarm"""
    def __init__(self, name: str, alphax: float, alphay: float, target: Target):
        self.name = name
        self.alphax = alphax
        self.betax = self._c_ampl(alphax)
        self.alphay = alphay
        self.betay = self._c_ampl(alphay)
        self.delta = self._reward(target, self.betax, self.betay)
        self.gamma = self._c_ampl(self.delta)
    
    @staticmethod
    def _c_ampl(ampl: float) -> float:
        """Compute complementary amplitude"""
        return np.sqrt(1 - ampl**2)
    
    @staticmethod
    def _reward(target: Target, betax: float, betay: float) -> float:
        """Calculate reward based on distance to target"""
        return 1 - (((target.x - betax)**2 + (target.y - betay)**2)**0.5) / np.sqrt(2)
    
    def update_position(self, new_alphax: float, new_alphay: float, target: Target):
        """Update robot position and recalculate reward"""
        self.alphax = new_alphax
        self.alphay = new_alphay
        self.betax = self._c_ampl(new_alphax)
        self.betay = self._c_ampl(new_alphay)
        self.delta = self._reward(target, self.betax, self.betay)
        self.gamma = self._c_ampl(self.delta)


class GroverPlanner:
    """Quantum path planner using Grover's algorithm"""
    def __init__(self, obstacles: List[Tuple]):
        self.simulator = Aer.get_backend("aer_simulator")
        self.obstacles = obstacles
    
    def XOR(self, qc, a, b, output):
        qc.cx(a, output)
        qc.cx(b, output)
    
    def encode_cell(self, qc, cell, qubits):
        for idx, bit in enumerate(cell):
            if int(bit):
                qc.x(qubits[idx])
    
    def encode_target(self, qc, target_cell, qubits):
        for idx, bit in enumerate(target_cell):
            if not int(bit):
                qc.x(qubits[idx])
    
    def obstacles_in_map(self, target_cell):
        return any(target_cell[1] == macro_cell for macro_cell in list(map(lambda ob_t: ob_t[1], self.obstacles)))
    
    def diffuser(self, nqubits):
        diff = QuantumCircuit(nqubits)
        for qubit in range(nqubits):
            diff.h(qubit)
        for qubit in range(nqubits):
            diff.x(qubit)
        diff.h(nqubits - 1)
        diff.mcx(list(range(nqubits - 1)), nqubits - 1, mode='noancilla')
        diff.h(nqubits - 1)
        for qubit in range(nqubits):
            diff.x(qubit)
        for qubit in range(nqubits):
            diff.h(qubit)
        U_s = diff.to_gate()
        U_s.name = "U$_\\psi$"
        return U_s
    
    def half_oracle(self, num_of_M, obstacles_in_map, quantum_registers):
        if num_of_M == 1:
            ho = QuantumCircuit(*quantum_registers)
            pos_qubits, mov_qubits, c, output = quantum_registers
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])
        elif num_of_M == 2:
            ho = QuantumCircuit(*quantum_registers)
            if obstacles_in_map:
                pos_qubits, mov_qubits, c, o, caux, out, output = quantum_registers
            else:
                pos_qubits, mov_qubits, c, output = quantum_registers
            
            self.XOR(ho, pos_qubits[1], mov_qubits[0], c[0])
            ho.x(c[0])
            self.XOR(ho, pos_qubits[0], mov_qubits[0], c[1])
            
            if obstacles_in_map:
                for obstacle in self.obstacles:
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])
                    ho.ccx(c[0], c[1], o)
                    self.encode_cell(ho, obstacle[0], [c[0], c[1]])
            
            self.XOR(ho, c[1], mov_qubits[1], c[2])
            ho.x(c[2])
            self.XOR(ho, c[0], mov_qubits[1], c[3])
            
            if obstacles_in_map:
                ho.ccx(c[2], o, caux[0])
                ho.x(caux[0])
                ho.ccx(c[3], o, caux[1])
                ho.x(caux[1])
                ho.x(o)
                ho.ccx(pos_qubits[0], o, caux[2])
                ho.x(caux[2])
                ho.ccx(pos_qubits[1], o, caux[3])
                ho.x(caux[3])
                ho.ccx(caux[0], caux[2], out[0])
                ho.x(out[0])
                ho.ccx(caux[1], caux[3], out[1])
                ho.x(out[1])
        
        HalfOracle = ho.to_gate()
        HalfOracle.name = "HO"
        return HalfOracle
    
    def oracle(self, num_of_M, target_cell, m_type, obstacles_in_map, qc, quantum_registers):
        halforacle = self.half_oracle(num_of_M, obstacles_in_map, quantum_registers)
        additional_regs = []
        
        if obstacles_in_map:
            pos_qubits, mov_qubits, c, *additional_regs, output = quantum_registers
        else:
            pos_qubits, mov_qubits, c, output = quantum_registers
        
        qc.append(
            halforacle,
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + ((additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:]) if additional_regs else [])
            + output[:],
        )
        
        test_qubits = c[-2:] if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map) else additional_regs[2]
        tcell_idx = 0 if m_type == "micro" else 1
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])
        qc.ccx(test_qubits[0], test_qubits[1], output)
        self.encode_target(qc, target_cell[tcell_idx], [test_qubits[0], test_qubits[1]])
        
        qc.append(
            halforacle.inverse(),
            pos_qubits[:]
            + mov_qubits[:]
            + c[:]
            + ((additional_regs[0][:] + additional_regs[1][:] + additional_regs[2][:]) if additional_regs else [])
            + output[:],
        )
    
    def find_path(self, start_cell, target_cell, m_type):
        cell_idx = 0 if m_type == "micro" else 1
        num_of_M = manhattan(start_cell[cell_idx], target_cell[cell_idx])
        
        pos_qubits = QuantumRegister(2, name="r")
        mov_qubits = QuantumRegister(num_of_M, name="m")
        c = QuantumRegister(2 * num_of_M, name="c")
        o = QuantumRegister(1, name="o")
        caux = QuantumRegister(4, name="caux")
        out = QuantumRegister(2, name="out")
        output = QuantumRegister(1, name="q")
        cbits = ClassicalRegister(2 + num_of_M, name="cbits")
        
        obstacles_in_map = self.obstacles_in_map(target_cell) and m_type == "micro"
        
        base_registers = [pos_qubits, mov_qubits, c, output, cbits]
        additional_registers = base_registers.copy()
        additional_registers[3:3] = [o, caux, out]
        quantum_registers = (
            base_registers if num_of_M == 1 or (num_of_M == 2 and not obstacles_in_map)
            else additional_registers
        )
        
        qc = QuantumCircuit(*quantum_registers)
        qc.h(pos_qubits)
        qc.h(mov_qubits)
        qc.initialize([1, -1] / np.sqrt(2), output)
        
        num_iterations = 1
        for _ in range(num_iterations):
            self.oracle(num_of_M, target_cell, m_type, obstacles_in_map, qc, quantum_registers[:-1])
            diff_qubits = [0, 1, 2] if num_of_M == 1 else list(range(2 + num_of_M))
            qc.append(self.diffuser(len(diff_qubits)), diff_qubits)
        
        qc.measure(pos_qubits, cbits[0:2])
        qc.measure(mov_qubits, cbits[2:(2 + num_of_M)])
        
        transpiled_qc = transpile(qc, self.simulator)
        result_counts = self.simulator.run(transpiled_qc).result().get_counts()
        
        return [k for k in result_counts.keys() if k[-2:] == start_cell[cell_idx][len(start_cell[cell_idx]):: -1]]


def find_centroids(dim_map: int) -> Dict:
    """Generate centroids for the map grid"""
    unit_displacement = 1 / (dim_map * 2)
    centroids_coords = []
    for i in range(dim_map, 0, -1):
        row = []
        for j in range(dim_map):
            row.append((j * unit_displacement * 2 + unit_displacement, 
                       i * unit_displacement * 2 - unit_displacement))
        centroids_coords.append(row)
    
    centroids_dict = {
        t: (f"{str(i%2)}{str(j%2)}", f"{0 if t[1] > 0.5 else 1}{0 if t[0] < 0.5 else 1}")
        for i, row in enumerate(centroids_coords) for j, t in enumerate(row)
    }
    return centroids_dict


def find_closest_centroid(centroids: Dict, x: float, y: float) -> Tuple:
    """Find the closest centroid to a given position"""
    centroids_coords = np.array(list(centroids.keys()))
    distances = np.linalg.norm(centroids_coords - np.array([x, y]), axis=1)
    return centroids[tuple(centroids_coords[np.argmin(distances)])]


def manhattan(start_cell: str, target_cell: str) -> int:
    """Calculate Manhattan distance between cells"""
    start_cell_v = [int(b) for b in start_cell]
    target_cell_v = [int(b) for b in target_cell]
    return int(cityblock(start_cell_v, target_cell_v).item())


def eval_displacement(old_centroid: Tuple, new_centroid: Tuple) -> List[int]:
    """Calculate displacement between centroids"""
    old_c_micro, old_c_macro = old_centroid
    new_c_micro, new_c_macro = new_centroid
    return [manhattan(old_c_micro, new_c_micro), manhattan(old_c_macro, new_c_macro)]


def compute_path_to_position(robot: Robot, new_alphas: List[float], 
                            centroids: Dict, grover_planner: GroverPlanner,
                            target: Target, verbose: bool = False):
    """Compute and execute path for robot to new position"""
    old_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)
    robot.update_position(new_alphas[0], new_alphas[1], target)
    new_centroid = find_closest_centroid(centroids, robot.betax, robot.betay)
    displacement = eval_displacement(old_centroid, new_centroid)
    
    if displacement[0] == 0 and displacement[1] != 0:
        if verbose:
            print(f"{robot.name}: macro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="macro")
        if verbose:
            print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")
    elif displacement[0] != 0 and displacement[1] == 0:
        if verbose:
            print(f"{robot.name}: micro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="micro")
        if verbose:
            print(f"{robot.name} evaluated path: {evaluated_path[0][:-2]}")
    elif displacement[0] != 0 and displacement[1] != 0:
        if verbose:
            print(f"{robot.name}: macro-micro")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="macro")
        if verbose:
            print(f"{robot.name} evaluated path (macro): {evaluated_path[0][:-2]}")
        evaluated_path = grover_planner.find_path(old_centroid, new_centroid, m_type="micro")
        if verbose:
            print(f"{robot.name} evaluated path (micro): {evaluated_path[0][:-2]}")


def is_position_on_obstacle(new_alphas: List[float], centroids: Dict, 
                           obstacles_centroids: Dict) -> bool:
    """Check if position collides with obstacle"""
    c_ampl = lambda a: np.sqrt(1 - a**2)
    new_pos_centroid = find_closest_centroid(centroids, c_ampl(new_alphas[0]), c_ampl(new_alphas[1]))
    return new_pos_centroid in obstacles_centroids.values()


def informed_alpha_reshuffling(robots: List[Robot], target: Target) -> List[float]:
    """Generate new position within search rectangle"""
    c_ampl = lambda a: np.sqrt(1 - a**2)
    centroids = find_centroids(4)
    
    farthest_robot = min(robots, key=attrgetter('delta'))
    f_centroid = find_closest_centroid(centroids, farthest_robot.betax, farthest_robot.betay)
    f_coords = list(list(centroids.keys())[list(centroids.values()).index(f_centroid)])
    
    if f_coords[0] == target.x:
        f_coords[0] = f_coords[1]
    elif f_coords[1] == target.y:
        f_coords[1] = f_coords[0]
    
    return [c_ampl(np.random.uniform(f_coords[0], target.x)), 
            c_ampl(np.random.uniform(f_coords[1], target.y))]


def calculate_avg_reward(robots: List[Robot]) -> float:
    """Calculate average reward across all robots"""
    return np.mean([robot.delta for robot in robots])


def run_quantum_swarm_experiment(
    num_robots: int = 10,
    obstacle_positions: List[Tuple[float, float]] = None,
    target_position: Tuple[float, float] = (0.875, 0.125),
    delta_threshold: float = 0.8,
    starting_position: Tuple[float, float] = (0.125, 0.875),
    starting_radius: float = 0.05,
    num_reshuffles: int = 2,
    num_gate_applications: int = 2,
    verbose: bool = False,
    plot: bool = False,
    plot_all_steps: bool = False  # NEW: Plot after every position change
) -> Dict:
    """
    Run a complete quantum swarm robotics experiment.
    
    Parameters:
    -----------
    num_robots : int
        Number of robots in the swarm
    obstacle_positions : List[Tuple[float, float]]
        List of (x, y) coordinates for obstacles
    target_position : Tuple[float, float]
        (x, y) coordinates of the target
    delta_threshold : float
        Reward threshold for triggering reshuffle
    starting_position : Tuple[float, float]
        Center of initial robot cluster
    starting_radius : float
        Radius of initial robot distribution
    num_reshuffles : int
        Number of reshuffle operations to perform
    num_gate_applications : int
        Number of times to apply the quantum gate
    verbose : bool
        Print detailed progress information
    plot : bool
        Generate visualization plots
    plot_all_steps : bool
        Plot grid after every position change (NEW)
    
    Returns:
    --------
    Dict containing experimental results and reward history
    """
    
    # Setup
    if obstacle_positions is None:
        obstacle_positions = [(0.125, 0.125), (0.875, 0.375), (0.625, 0.125)]
    
    target = Target("T", target_position[0], target_position[1])
    obstacles = [Obstacle(f"O{i+1}", pos[0], pos[1]) for i, pos in enumerate(obstacle_positions)]
    
    # Initialize robots
    c_ampl = lambda a: np.sqrt(1 - a**2)
    a_x = c_ampl(starting_position[0] - starting_radius)
    a_y = c_ampl(starting_position[0] + starting_radius)
    b_x = c_ampl(starting_position[1] - starting_radius)
    b_y = c_ampl(starting_position[1] + starting_radius)
    
    robots = []
    for i in range(num_robots):
        robot = Robot(f'R{i}', 
                     np.random.uniform(a_x, a_y), 
                     np.random.uniform(b_x, b_y),
                     target)
        robots.append(robot)
    
    # Setup environment
    centroids = find_centroids(4)
    obstacles_centroids = {
        coord: v for (coord, v) in centroids.items() 
        for o in obstacles if coord == (o.x, o.y)
    }
    grover_planner = GroverPlanner(list(obstacles_centroids.values()))
    
    # Track rewards at different stages
    rewards_history = {
        't0_initial': calculate_avg_reward(robots),
        'reshuffles': [],
        'gates': [],
        'final': None
    }
    
    if verbose:
        print(f"Initial average reward: {rewards_history['t0_initial']:.4f}")
    
    # Plot initial configuration
    if plot or plot_all_steps:
        plot_experiment_results(robots, target, obstacles, 
                               title=f"Initial Configuration - Avg Reward: {rewards_history['t0_initial']:.4f}")
    
    # Main experiment loop
    for reshuffle_idx in range(num_reshuffles):
        # Check if reshuffle needed
        if all(r.delta < delta_threshold for r in robots):
            if verbose:
                print(f"\n=== Reshuffle {reshuffle_idx + 1} ===")
            
            for robot in robots:
                new_alphas = informed_alpha_reshuffling(robots, target)
                while is_position_on_obstacle(new_alphas, centroids, obstacles_centroids):
                    new_alphas = informed_alpha_reshuffling(robots, target)
                compute_path_to_position(robot, new_alphas, centroids, grover_planner, target, verbose)
            
            rewards_history['reshuffles'].append(calculate_avg_reward(robots))
            if verbose:
                print(f"Average reward after reshuffle: {rewards_history['reshuffles'][-1]:.4f}")
            
            # Plot after reshuffle
            if plot_all_steps:
                plot_experiment_results(robots, target, obstacles,
                                       title=f"After Reshuffle {reshuffle_idx + 1} - Avg Reward: {rewards_history['reshuffles'][-1]:.4f}")
        
        # Apply quantum gate
        if reshuffle_idx < num_gate_applications:
            if verbose:
                print(f"\n=== Quantum Gate {reshuffle_idx + 1} ===")
            
            closest_robot = max(robots, key=attrgetter('delta'))
            
            if verbose:
                print(f"Robot entering gate: {closest_robot.name} with reward {closest_robot.delta:.4f}")
            
            # Quantum circuit setup
            q = QuantumRegister(5, 'q')
            m2 = ClassicalRegister(1, 'c1')
            m3 = ClassicalRegister(1, 'c2')
            m4 = ClassicalRegister(1, 'c3')
            qc = QuantumCircuit(q, m2, m3, m4)
            
            # Initialize quantum state
            vector0 = np.array([closest_robot.alphax, closest_robot.betax])
            vector1 = np.array([closest_robot.alphay, closest_robot.betay])
            vector3 = np.array([closest_robot.gamma, closest_robot.delta])
            
            qc.initialize(vector0 / np.linalg.norm(vector0), q[0])
            qc.initialize(vector1 / np.linalg.norm(vector1), q[1])
            qc.initialize(vector3 / np.linalg.norm(vector3), q[2])
            
            # Apply quantum operations
            qc.barrier(q)
            qc.ccx(q[0], q[2], q[3])
            qc.ccx(q[1], q[2], q[4])
            qc.x(q[2])
            qc.ch(q[2], q[3])
            qc.ch(q[2], q[4])
            qc.x(q[2])
            qc.barrier(q)
            
            # Measure
            qc.measure(q[2], m2[0])
            qc.measure(q[3], m3[0])
            qc.measure(q[4], m4[0])
            
            # Simulate
            aer_sim = Aer.get_backend("aer_simulator")
            transpiled_qc = transpile(qc, aer_sim)
            result = aer_sim.run(transpiled_qc).result()
            counts = result.get_counts()
            
            if verbose:
                print(f"Quantum measurement counts: {counts}")
            
            # Extract outcomes
            num_most_prob_states = min(4, len(counts))
            most_prob_dict = dict(nlargest(num_most_prob_states, counts.items(), key=itemgetter(1)))
            
            def eval_outcome(most_prob_dict, n_outcome):
                mapped_weights = [int(res[n_outcome * 2]) * most_prob_dict[res] for res in most_prob_dict]
                return sum(mapped_weights) / sum(most_prob_dict.values())
            
            outcome0, outcome1 = eval_outcome(most_prob_dict, 0), eval_outcome(most_prob_dict, 1)
            
            if verbose:
                print(f"Quantum outcomes: outcome0={outcome0:.4f}, outcome1={outcome1:.4f}")
            
            # Update other robots based on quantum outcomes
            if verbose:
                print("Updating other robots based on quantum outcomes...")
            
            for robot in robots:
                if robot.delta != closest_robot.delta or all(r.delta == robots[0].delta for r in robots):
                    compute_path_to_position(robot, [c_ampl(outcome1), c_ampl(outcome0)], 
                                           centroids, grover_planner, target, verbose)
            
            # Plot after quantum gate outcomes
            if plot_all_steps:
                avg_reward_after_outcomes = calculate_avg_reward(robots)
                plot_experiment_results(robots, target, obstacles,
                                       title=f"After Gate {reshuffle_idx + 1} Outcomes - Avg Reward: {avg_reward_after_outcomes:.4f}")
            
            # Cluster around best robot
            if verbose:
                print("Clustering robots around most successful robot...")
            
            max_robot = max(robots, key=attrgetter('delta'))
            
            if verbose:
                print(f"Most successful robot: {max_robot.name} with reward {max_robot.delta:.4f}")
            
            for robot in robots:
                if robot != max_robot:
                    compute_path_to_position(
                        robot,
                        [max_robot.alphax + np.random.uniform(0, 0.01),
                         max_robot.alphay + np.random.uniform(0, 0.01)],
                        centroids, grover_planner, target, verbose
                    )
            
            rewards_history['gates'].append(calculate_avg_reward(robots))
            if verbose:
                print(f"Average reward after gate clustering: {rewards_history['gates'][-1]:.4f}")
            
            # Plot after clustering
            if plot_all_steps:
                plot_experiment_results(robots, target, obstacles,
                                       title=f"After Gate {reshuffle_idx + 1} Clustering - Avg Reward: {rewards_history['gates'][-1]:.4f}")
    
    # Final reward
    rewards_history['final'] = calculate_avg_reward(robots)
    
    if verbose:
        print(f"\n=== Experiment Complete ===")
        print(f"Final average reward: {rewards_history['final']:.4f}")
        print(f"Total obstacles: {len(obstacles)}")
    
    # Plot final configuration
    if plot or plot_all_steps:
        plot_experiment_results(robots, target, obstacles,
                               title=f"Final Configuration - Avg Reward: {rewards_history['final']:.4f}")
    
    return {
        'robots': robots,
        'target': target,
        'obstacles': obstacles,
        'rewards_history': rewards_history,
        'num_robots': num_robots,
        'delta_threshold': delta_threshold
    }


def plot_experiment_results(robots: List[Robot], target: Target, 
                           obstacles: List[Obstacle], title: str = "Experiment Results",
                           show_labels: bool = True):
    """Visualize the final state of the experiment"""
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_xticks(np.arange(0, 1, 0.25))
    ax.set_yticks(np.arange(0, 1, 0.25))
    plt.grid(True, alpha=0.3, linewidth=1.5)
    
    # Add minor grid for cells
    ax.set_xticks(np.arange(0, 1, 0.125), minor=True)
    ax.set_yticks(np.arange(0, 1, 0.125), minor=True)
    plt.grid(which='minor', alpha=0.2, linestyle=':', linewidth=0.5)
    
    for i, robot in enumerate(robots):
        # Color gradient based on reward (green = high reward, yellow = low reward)
        reward_color = plt.cm.RdYlGn(robot.delta)
        plt.scatter(robot.betax, robot.betay, s=300, marker='o', 
                   color=reward_color, alpha=0.8, edgecolors='black', linewidth=1.5)
        if show_labels:
            plt.text(robot.betax + 0.02, robot.betay + 0.02, robot.name, 
                    fontsize=8, fontweight='bold')
    
    plt.scatter(target.x, target.y, s=500, marker='*', color='turquoise', 
               label='Target', edgecolors='black', linewidth=2, zorder=10)
    
    for obstacle in obstacles:
        plt.scatter(obstacle.x, obstacle.y, s=450, marker='p', color='red', 
                   alpha=0.7, label='Obstacle' if obstacle == obstacles[0] else '',
                   edgecolors='darkred', linewidth=2)
        if show_labels:
            plt.text(obstacle.x + 0.02, obstacle.y + 0.02, obstacle.name, 
                    fontsize=8, color='darkred', fontweight='bold')
    
    plt.axis([0, 1, 0, 1])
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel('X Position', fontsize=12)
    plt.ylabel('Y Position', fontsize=12)
    plt.legend(fontsize=10)
    
    # Add colorbar for robot rewards
    sm = plt.cm.ScalarMappable(cmap=plt.cm.RdYlGn, 
                               norm=plt.Normalize(vmin=0, vmax=1))
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax, label='Robot Reward')
    
    plt.tight_layout()
    plt.show()


def get_mean_rewards_at_stages(rewards_history: Dict) -> Dict[str, float]:
    """
    Extract mean rewards at different stages of the experiment.
    
    Parameters:
    -----------
    rewards_history : Dict
        The rewards_history from run_quantum_swarm_experiment
    
    Returns:
    --------
    Dict with keys 't0', 't1', 't2', 't3' representing:
        t0: Initial state
        t1: After first reshuffle
        t2: After first logic gate
        t3: Final state
    """
    result = {
        't0': rewards_history['t0_initial'],
        't1': rewards_history['reshuffles'][0] if rewards_history['reshuffles'] else None,
        't2': rewards_history['gates'][0] if rewards_history['gates'] else None,
        't3': rewards_history['final']
    }
    
    return result


def run_multiple_experiments(
    num_experiments: int = 10,
    **experiment_params
) -> List[Dict]:
    """
    Run multiple experiments and collect results.
    
    Parameters:
    -----------
    num_experiments : int
        Number of experiments to run
    **experiment_params : 
        Parameters to pass to run_quantum_swarm_experiment
    
    Returns:
    --------
    List of experiment results
    """
    results = []
    
    for i in range(num_experiments):
        print(f"\n{'='*50}")
        print(f"Running Experiment {i + 1}/{num_experiments}")
        print(f"{'='*50}")
        
        result = run_quantum_swarm_experiment(
            verbose=experiment_params.get('verbose', False),
            plot=experiment_params.get('plot', False),
            **{k: v for k, v in experiment_params.items() if k not in ['verbose', 'plot']}
        )
        
        results.append(result)
        
        # Print stage rewards
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        print(f"\nStage Rewards:")
        for stage, reward in stage_rewards.items():
            print(f"  {stage}: {reward:.4f}" if reward is not None else f"  {stage}: N/A")
    
    return results


def aggregate_experiment_statistics(results: List[Dict]) -> Dict:
    """
    Aggregate statistics across multiple experiments.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results from run_multiple_experiments
    
    Returns:
    --------
    Dictionary containing aggregated statistics
    """
    all_stage_rewards = {
        't0': [],
        't1': [],
        't2': [],
        't3': []
    }
    
    for result in results:
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        for stage, reward in stage_rewards.items():
            if reward is not None:
                all_stage_rewards[stage].append(reward)
    
    statistics = {}
    for stage, rewards in all_stage_rewards.items():
        if rewards:
            statistics[stage] = {
                'mean': np.mean(rewards),
                'std': np.std(rewards),
                'min': np.min(rewards),
                'max': np.max(rewards),
                'median': np.median(rewards)
            }
        else:
            statistics[stage] = None
    
    return statistics


def plot_reward_progression(results: List[Dict], title: str = "Reward Progression Across Experiments"):
    """
    Plot reward progression across multiple experiments.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results
    title : str
        Plot title
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Individual experiment trajectories
    for idx, result in enumerate(results):
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        stages = ['t0', 't1', 't2', 't3']
        rewards = [stage_rewards.get(s) for s in stages]
        
        # Filter out None values
        valid_stages = [s for s, r in zip(stages, rewards) if r is not None]
        valid_rewards = [r for r in rewards if r is not None]
        
        ax1.plot(valid_stages, valid_rewards, marker='o', alpha=0.5, label=f'Exp {idx+1}')
    
    ax1.set_xlabel('Stage')
    ax1.set_ylabel('Average Reward')
    ax1.set_title('Individual Experiment Trajectories')
    ax1.grid(True, alpha=0.3)
    ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # Aggregated statistics
    stats = aggregate_experiment_statistics(results)
    stages = ['t0', 't1', 't2', 't3']
    means = [stats[s]['mean'] if stats[s] else None for s in stages]
    stds = [stats[s]['std'] if stats[s] else 0 for s in stages]
    
    valid_stages = [s for s, m in zip(stages, means) if m is not None]
    valid_means = [m for m in means if m is not None]
    valid_stds = [std for m, std in zip(means, stds) if m is not None]
    
    ax2.errorbar(valid_stages, valid_means, yerr=valid_stds, 
                 marker='o', markersize=10, linewidth=2, capsize=5,
                 label='Mean ± Std')
    ax2.fill_between(range(len(valid_means)), 
                      np.array(valid_means) - np.array(valid_stds),
                      np.array(valid_means) + np.array(valid_stds),
                      alpha=0.3)
    
    ax2.set_xlabel('Stage')
    ax2.set_ylabel('Average Reward')
    ax2.set_title('Aggregated Statistics (Mean ± Std)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()


def compare_experiment_configurations(configurations: List[Dict]) -> None:
    """
    Compare results across different experiment configurations.
    
    Parameters:
    -----------
    configurations : List[Dict]
        List of configuration dictionaries, each containing:
        - 'name': configuration name
        - 'params': parameters for run_multiple_experiments
        - 'num_experiments': number of experiments to run
    
    Example:
    --------
    configs = [
        {
            'name': 'Few Obstacles',
            'params': {'obstacle_positions': [(0.125, 0.125)]},
            'num_experiments': 5
        },
        {
            'name': 'Many Obstacles',
            'params': {'obstacle_positions': [(0.125, 0.125), (0.625, 0.625), (0.875, 0.375)]},
            'num_experiments': 5
        }
    ]
    compare_experiment_configurations(configs)
    """
    all_results = {}
    
    for config in configurations:
        print(f"\n{'='*60}")
        print(f"Testing Configuration: {config['name']}")
        print(f"{'='*60}")
        
        results = run_multiple_experiments(
            num_experiments=config['num_experiments'],
            **config['params']
        )
        
        all_results[config['name']] = results
        
        # Print summary statistics
        stats = aggregate_experiment_statistics(results)
        print(f"\n{config['name']} - Summary Statistics:")
        for stage, stat in stats.items():
            if stat:
                print(f"  {stage}: {stat['mean']:.4f} ± {stat['std']:.4f} "
                      f"(min: {stat['min']:.4f}, max: {stat['max']:.4f})")
    
    # Comparative visualization
    fig, axes = plt.subplots(1, len(configurations), figsize=(6*len(configurations), 5))
    if len(configurations) == 1:
        axes = [axes]
    
    for idx, (config_name, results) in enumerate(all_results.items()):
        stats = aggregate_experiment_statistics(results)
        stages = ['t0', 't1', 't2', 't3']
        means = [stats[s]['mean'] if stats[s] else None for s in stages]
        stds = [stats[s]['std'] if stats[s] else 0 for s in stages]
        
        valid_stages = [s for s, m in zip(stages, means) if m is not None]
        valid_means = [m for m in means if m is not None]
        valid_stds = [std for m, std in zip(means, stds) if m is not None]
        
        axes[idx].errorbar(valid_stages, valid_means, yerr=valid_stds,
                          marker='o', markersize=10, linewidth=2, capsize=5)
        axes[idx].fill_between(range(len(valid_means)),
                               np.array(valid_means) - np.array(valid_stds),
                               np.array(valid_means) + np.array(valid_stds),
                               alpha=0.3)
        axes[idx].set_xlabel('Stage')
        axes[idx].set_ylabel('Average Reward')
        axes[idx].set_title(config_name)
        axes[idx].grid(True, alpha=0.3)
        axes[idx].set_ylim([0, 1])
    
    plt.tight_layout()
    plt.show()
    
    return all_results


def export_results_to_csv(results: List[Dict], filename: str = "experiment_results.csv"):
    """
    Export experiment results to CSV file.
    
    Parameters:
    -----------
    results : List[Dict]
        List of experiment results
    filename : str
        Output CSV filename
    """
    import csv
    
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['experiment_id', 'num_robots', 'num_obstacles', 'delta_threshold',
                     't0_reward', 't1_reward', 't2_reward', 't3_reward',
                     'improvement_t0_to_t3']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for idx, result in enumerate(results):
            stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
            
            row = {
                'experiment_id': idx + 1,
                'num_robots': result['num_robots'],
                'num_obstacles': len(result['obstacles']),
                'delta_threshold': result['delta_threshold'],
                't0_reward': stage_rewards.get('t0'),
                't1_reward': stage_rewards.get('t1'),
                't2_reward': stage_rewards.get('t2'),
                't3_reward': stage_rewards.get('t3'),
                'improvement_t0_to_t3': (stage_rewards.get('t3', 0) - stage_rewards.get('t0', 0))
            }
            writer.writerow(row)
    
    print(f"Results exported to {filename}")


# Example usage demonstrations:
if __name__ == "__main__":
    
    print("="*60)
    print("EXAMPLE 1: Single Experiment with Step-by-Step Visualization")
    print("="*60)
    
    result = run_quantum_swarm_experiment(
        num_robots=10,
        obstacle_positions=[(0.125, 0.125), (0.875, 0.375), (0.625, 0.125)],
        target_position=(0.875, 0.125),
        delta_threshold=0.8,
        starting_position=(0.125, 0.875),
        starting_radius=0.05,
        num_reshuffles=2,
        num_gate_applications=2,
        verbose=True,
        plot=False,
        plot_all_steps=True  # This will show grid after every position change
    )
    
    stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
    print("\nMean Rewards at Different Stages:")
    for stage, reward in stage_rewards.items():
        print(f"  {stage}: {reward:.4f}" if reward else f"  {stage}: N/A")
    
    
    print("\n" + "="*60)
    print("EXAMPLE 2: Multiple Experiments with Final Plots Only")
    print("="*60)
    
    experiments = run_multiple_experiments(
        num_experiments=3,
        num_robots=8,
        delta_threshold=0.75,
        obstacle_positions=[(0.125, 0.125), (0.625, 0.625)],
        verbose=False,
        plot=False
    )
    
    # Show aggregated statistics
    stats = aggregate_experiment_statistics(experiments)
    print("\nAggregated Statistics Across Experiments:")
    for stage, stat in stats.items():
        if stat:
            print(f"\n{stage}:")
            print(f"  Mean: {stat['mean']:.4f}")
            print(f"  Std:  {stat['std']:.4f}")
            print(f"  Min:  {stat['min']:.4f}")
            print(f"  Max:  {stat['max']:.4f}")
    
    # Plot progression
    plot_reward_progression(experiments, "Reward Progression - 8 Robots, 2 Obstacles")
    
    
    print("\n" + "="*60)
    print("="*60)
    
    print("EXAMPLE 3: Comparing Different Configurations")
    configurations = [
        {
            'name': '5 Robots - Low Threshold',
            'params': {
                'num_robots': 5,
                'delta_threshold': 0.6,
                'obstacle_positions': [(0.125, 0.125), (0.625, 0.625)]
            },
            'num_experiments': 3
        },
        {
            'name': '15 Robots - High Threshold',
            'params': {
                'num_robots': 15,
                'delta_threshold': 0.9,
                'obstacle_positions': [(0.125, 0.125), (0.625, 0.625)]
            },
            'num_experiments': 3
        }
    ]
    
    comparison_results = compare_experiment_configurations(configurations)
    
    
    print("\n" + "="*60)
    print("EXAMPLE 4: Export Results to CSV")
    print("="*60)
    
    export_results_to_csv(experiments, "quantum_swarm_results.csv")
```

```python
# -*- coding: utf-8 -*-
"""
Comprehensive Quantum Swarm Comparison Analysis
- Compare different swarm sizes
- Compare different obstacle densities
- Test with blocked paths
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from typing import List, Dict, Tuple
import seaborn as sns

# Assuming the main framework functions are imported
# from quantum_swarm_experiments import *

def generate_random_obstacles(num_obstacles: int, 
                             target_pos: Tuple[float, float],
                             min_distance: float = 0.2,
                             seed: int = None) -> List[Tuple[float, float]]:
    """
    Generate random obstacle positions avoiding target location.
    
    Parameters:
    -----------
    num_obstacles : int
        Number of obstacles to generate
    target_pos : Tuple[float, float]
        Target position to avoid
    min_distance : float
        Minimum distance from target
    seed : int
        Random seed for reproducibility
    
    Returns:
    --------
    List of (x, y) obstacle coordinates
    """
    if seed is not None:
        np.random.seed(seed)
    
    obstacles = []
    # Possible cell centers in the 4x4 grid
    possible_positions = []
    for i in range(4):
        for j in range(4):
            x = 0.125 + i * 0.25
            y = 0.125 + j * 0.25
            # Check distance from target
            dist = np.sqrt((x - target_pos[0])**2 + (y - target_pos[1])**2)
            if dist >= min_distance:
                possible_positions.append((x, y))
    
    # Randomly sample obstacles
    if num_obstacles > len(possible_positions):
        print(f"Warning: Requested {num_obstacles} obstacles but only {len(possible_positions)} valid positions available")
        num_obstacles = len(possible_positions)
    
    selected_indices = np.random.choice(len(possible_positions), num_obstacles, replace=False)
    obstacles = [possible_positions[i] for i in selected_indices]
    
    return obstacles


def block_path_to_target(target_pos: Tuple[float, float],
                         starting_pos: Tuple[float, float] = (0.125, 0.875),
                         num_blocking_obstacles: int = 5) -> List[Tuple[float, float]]:
    """
    Create obstacles that block direct paths to the target.
    
    Parameters:
    -----------
    target_pos : Tuple[float, float]
        Target position
    starting_pos : Tuple[float, float]
        Starting position of robots
    num_blocking_obstacles : int
        Number of obstacles to place blocking the path
    
    Returns:
    --------
    List of obstacle positions blocking the path
    """
    obstacles = []
    
    # Create a barrier around the target
    target_x, target_y = target_pos
    
    # Define cells around target to block
    blocking_positions = [
        (target_x - 0.25, target_y),      # Left
        (target_x, target_y + 0.25),      # Top
        (target_x - 0.25, target_y + 0.25),  # Top-left
        (target_x + 0.25, target_y),      # Right (if valid)
        (target_x, target_y - 0.25),      # Bottom (if valid)
        (target_x - 0.25, target_y - 0.25),  # Bottom-left
        (target_x - 0.5, target_y),       # Further left
        (target_x, target_y + 0.5),       # Further top
    ]
    
    # Filter valid positions (within bounds)
    valid_blocking = [(x, y) for x, y in blocking_positions 
                     if 0.125 <= x <= 0.875 and 0.125 <= y <= 0.875
                     and (x, y) != target_pos]
    
    # Randomly select from valid positions
    num_to_select = min(num_blocking_obstacles, len(valid_blocking))
    selected_indices = np.random.choice(len(valid_blocking), num_to_select, replace=False)
    obstacles = [valid_blocking[i] for i in selected_indices]
    
    return obstacles


def experiment_1_swarm_size_comparison(
    swarm_sizes: List[int] = [2, 3, 5, 10],
    num_experiments_per_size: int = 10,
    obstacle_positions: List[Tuple[float, float]] = None,
    target_position: Tuple[float, float] = (0.875, 0.125),
    delta_threshold: float = 0.8
) -> Dict:
    """
    Compare average rewards for swarms of different sizes.
    
    Parameters:
    -----------
    swarm_sizes : List[int]
        List of swarm sizes to test
    num_experiments_per_size : int
        Number of experiments per swarm size
    obstacle_positions : List[Tuple[float, float]]
        Fixed obstacle positions for all experiments
    target_position : Tuple[float, float]
        Target position
    delta_threshold : float
        Reward threshold
    
    Returns:
    --------
    Dictionary with results and statistics
    """
    print("\n" + "="*70)
    print("EXPERIMENT 1: Swarm Size Comparison")
    print(f"Testing swarm sizes: {swarm_sizes}")
    print(f"Experiments per size: {num_experiments_per_size}")
    print("="*70 + "\n")
    
    if obstacle_positions is None:
        obstacle_positions = [(0.125, 0.125), (0.625, 0.625)]
    
    all_results = {}
    summary_data = []
    
    for size in swarm_sizes:
        print(f"\n--- Testing swarm size: {size} robots ---")
        
        experiments = run_multiple_experiments(
            num_experiments=num_experiments_per_size,
            num_robots=size,
            obstacle_positions=obstacle_positions,
            target_position=target_position,
            delta_threshold=delta_threshold,
            verbose=False,
            plot=False,
            plot_all_steps=False
        )
        
        all_results[f"{size}_robots"] = experiments
        
        # Extract stage rewards for each experiment
        for exp_idx, exp in enumerate(experiments):
            stage_rewards = get_mean_rewards_at_stages(exp['rewards_history'])
            summary_data.append({
                'swarm_size': size,
                'experiment': exp_idx + 1,
                't0': stage_rewards['t0'],
                't1': stage_rewards['t1'],
                't2': stage_rewards['t2'],
                't3': stage_rewards['t3'],
                'improvement': stage_rewards['t3'] - stage_rewards['t0']
            })
        
        # Print summary for this size
        stats = aggregate_experiment_statistics(experiments)
        print(f"  t0 (Initial): {stats['t0']['mean']:.4f} ± {stats['t0']['std']:.4f}")
        print(f"  t3 (Final):   {stats['t3']['mean']:.4f} ± {stats['t3']['std']:.4f}")
        print(f"  Improvement:  {stats['t3']['mean'] - stats['t0']['mean']:.4f}")
    
    # Create DataFrame for analysis
    df = pd.DataFrame(summary_data)
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Box plot of final rewards by swarm size
    ax1 = axes[0, 0]
    df.boxplot(column='t3', by='swarm_size', ax=ax1)
    ax1.set_title('Final Reward Distribution by Swarm Size')
    ax1.set_xlabel('Number of Robots')
    ax1.set_ylabel('Final Reward (t3)')
    ax1.get_figure().suptitle('')
    
    # Plot 2: Average reward progression by swarm size
    ax2 = axes[0, 1]
    for size in swarm_sizes:
        size_data = df[df['swarm_size'] == size]
        stages = ['t0', 't1', 't2', 't3']
        means = [size_data[stage].mean() for stage in stages]
        stds = [size_data[stage].std() for stage in stages]
        ax2.errorbar(stages, means, yerr=stds, marker='o', label=f'{size} robots', capsize=5)
    ax2.set_title('Reward Progression by Swarm Size')
    ax2.set_xlabel('Stage')
    ax2.set_ylabel('Average Reward')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Improvement by swarm size
    ax3 = axes[1, 0]
    df.boxplot(column='improvement', by='swarm_size', ax=ax3)
    ax3.set_title('Reward Improvement by Swarm Size')
    ax3.set_xlabel('Number of Robots')
    ax3.set_ylabel('Improvement (t3 - t0)')
    ax3.get_figure().suptitle('')
    ax3.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    
    # Plot 4: Heatmap of average rewards at each stage
    ax4 = axes[1, 1]
    pivot_data = df.groupby('swarm_size')[['t0', 't1', 't2', 't3']].mean()
    sns.heatmap(pivot_data.T, annot=True, fmt='.3f', cmap='YlGnBu', ax=ax4)
    ax4.set_title('Average Rewards Heatmap')
    ax4.set_xlabel('Swarm Size')
    ax4.set_ylabel('Stage')
    
    plt.tight_layout()
    plt.savefig('swarm_size_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'results': all_results,
        'dataframe': df,
        'summary': df.groupby('swarm_size').describe()
    }


def experiment_2_obstacle_density_comparison(
    obstacle_counts: List[int] = [1, 2, 5, 10],
    num_experiments_per_density: int = 10,
    num_robots: int = 10,
    target_position: Tuple[float, float] = (0.875, 0.125),
    delta_threshold: float = 0.8,
    use_random_obstacles: bool = True
) -> Dict:
    """
    Compare average rewards with different numbers of obstacles.
    
    Parameters:
    -----------
    obstacle_counts : List[int]
        List of obstacle counts to test
    num_experiments_per_density : int
        Number of experiments per obstacle density
    num_robots : int
        Number of robots in swarm
    target_position : Tuple[float, float]
        Target position
    delta_threshold : float
        Reward threshold
    use_random_obstacles : bool
        If True, generate random obstacles for each experiment
    
    Returns:
    --------
    Dictionary with results and statistics
    """
    print("\n" + "="*70)
    print("EXPERIMENT 2: Obstacle Density Comparison")
    print(f"Testing obstacle counts: {obstacle_counts}")
    print(f"Experiments per density: {num_experiments_per_density}")
    print(f"Random obstacles: {use_random_obstacles}")
    print("="*70 + "\n")
    
    all_results = {}
    summary_data = []
    
    for num_obstacles in obstacle_counts:
        print(f"\n--- Testing with {num_obstacles} obstacle(s) ---")
        
        experiments = []
        for exp_idx in range(num_experiments_per_density):
            # Generate obstacles
            if use_random_obstacles:
                obstacles = generate_random_obstacles(
                    num_obstacles, 
                    target_position,
                    seed=exp_idx  # Different seed for each experiment
                )
            else:
                # Use fixed pattern
                if num_obstacles == 1:
                    obstacles = [(0.125, 0.125)]
                elif num_obstacles == 2:
                    obstacles = [(0.125, 0.125), (0.625, 0.625)]
                elif num_obstacles == 5:
                    obstacles = [(0.125, 0.125), (0.375, 0.375), (0.625, 0.625),
                               (0.125, 0.625), (0.625, 0.125)]
                else:
                    obstacles = generate_random_obstacles(num_obstacles, target_position, seed=exp_idx)
            
            result = run_quantum_swarm_experiment(
                num_robots=num_robots,
                obstacle_positions=obstacles,
                target_position=target_position,
                delta_threshold=delta_threshold,
                verbose=False,
                plot=False,
                plot_all_steps=False
            )
            
            experiments.append(result)
            
            # Extract stage rewards
            stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
            summary_data.append({
                'num_obstacles': num_obstacles,
                'experiment': exp_idx + 1,
                't0': stage_rewards['t0'],
                't1': stage_rewards['t1'],
                't2': stage_rewards['t2'],
                't3': stage_rewards['t3'],
                'improvement': stage_rewards['t3'] - stage_rewards['t0']
            })
        
        all_results[f"{num_obstacles}_obstacles"] = experiments
        
        # Print summary for this density
        stats = aggregate_experiment_statistics(experiments)
        print(f"  t0 (Initial): {stats['t0']['mean']:.4f} ± {stats['t0']['std']:.4f}")
        print(f"  t3 (Final):   {stats['t3']['mean']:.4f} ± {stats['t3']['std']:.4f}")
        print(f"  Improvement:  {stats['t3']['mean'] - stats['t0']['mean']:.4f}")
    
    # Create DataFrame for analysis
    df = pd.DataFrame(summary_data)
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Box plot of final rewards by obstacle count
    ax1 = axes[0, 0]
    df.boxplot(column='t3', by='num_obstacles', ax=ax1)
    ax1.set_title('Final Reward Distribution by Obstacle Count')
    ax1.set_xlabel('Number of Obstacles')
    ax1.set_ylabel('Final Reward (t3)')
    ax1.get_figure().suptitle('')
    
    # Plot 2: Average reward progression by obstacle count
    ax2 = axes[0, 1]
    for num_obs in obstacle_counts:
        obs_data = df[df['num_obstacles'] == num_obs]
        stages = ['t0', 't1', 't2', 't3']
        means = [obs_data[stage].mean() for stage in stages]
        stds = [obs_data[stage].std() for stage in stages]
        ax2.errorbar(stages, means, yerr=stds, marker='o', label=f'{num_obs} obstacles', capsize=5)
    ax2.set_title('Reward Progression by Obstacle Count')
    ax2.set_xlabel('Stage')
    ax2.set_ylabel('Average Reward')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Improvement by obstacle count
    ax3 = axes[1, 0]
    df.boxplot(column='improvement', by='num_obstacles', ax=ax3)
    ax3.set_title('Reward Improvement by Obstacle Count')
    ax3.set_xlabel('Number of Obstacles')
    ax3.set_ylabel('Improvement (t3 - t0)')
    ax3.get_figure().suptitle('')
    ax3.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    
    # Plot 4: Scatter plot showing relationship
    ax4 = axes[1, 1]
    for stage in ['t0', 't3']:
        stage_means = df.groupby('num_obstacles')[stage].mean()
        ax4.plot(stage_means.index, stage_means.values, marker='o', label=stage, linewidth=2)
    ax4.set_title('Reward vs Obstacle Density')
    ax4.set_xlabel('Number of Obstacles')
    ax4.set_ylabel('Average Reward')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('obstacle_density_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'results': all_results,
        'dataframe': df,
        'summary': df.groupby('num_obstacles').describe()
    }


def experiment_3_blocked_path_test(
    num_experiments: int = 10,
    num_robots: int = 10,
    target_position: Tuple[float, float] = (0.875, 0.125),
    starting_position: Tuple[float, float] = (0.125, 0.875),
    num_blocking_obstacles: int = 5,
    delta_threshold: float = 0.8
) -> Dict:
    """
    Test swarm performance when direct path to target is blocked.
    
    Parameters:
    -----------
    num_experiments : int
        Number of experiments to run
    num_robots : int
        Number of robots in swarm
    target_position : Tuple[float, float]
        Target position
    starting_position : Tuple[float, float]
        Starting position
    num_blocking_obstacles : int
        Number of obstacles blocking the path
    delta_threshold : float
        Reward threshold
    
    Returns:
    --------
    Dictionary with results and comparison to unblocked scenario
    """
    print("\n" + "="*70)
    print("EXPERIMENT 3: Blocked Path Test")
    print(f"Number of experiments: {num_experiments}")
    print(f"Blocking obstacles: {num_blocking_obstacles}")
    print("="*70 + "\n")
    
    # Experiment 3a: With blocked path
    print("--- Running experiments with BLOCKED path ---")
    blocked_experiments = []
    blocked_summary_data = []
    
    for exp_idx in range(num_experiments):
        obstacles = block_path_to_target(
            target_position, 
            starting_position,
            num_blocking_obstacles
        )
        
        result = run_quantum_swarm_experiment(
            num_robots=num_robots,
            obstacle_positions=obstacles,
            target_position=target_position,
            starting_position=starting_position,
            delta_threshold=delta_threshold,
            verbose=False,
            plot=False,
            plot_all_steps=False
        )
        
        blocked_experiments.append(result)
        
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        blocked_summary_data.append({
            'condition': 'blocked',
            'experiment': exp_idx + 1,
            't0': stage_rewards['t0'],
            't1': stage_rewards['t1'],
            't2': stage_rewards['t2'],
            't3': stage_rewards['t3'],
            'improvement': stage_rewards['t3'] - stage_rewards['t0']
        })
    
    # Experiment 3b: Without obstacles (control)
    print("\n--- Running experiments with NO obstacles (control) ---")
    unblocked_experiments = []
    unblocked_summary_data = []
    
    for exp_idx in range(num_experiments):
        result = run_quantum_swarm_experiment(
            num_robots=num_robots,
            obstacle_positions=[],  # No obstacles
            target_position=target_position,
            starting_position=starting_position,
            delta_threshold=delta_threshold,
            verbose=False,
            plot=False,
            plot_all_steps=False
        )
        
        unblocked_experiments.append(result)
        
        stage_rewards = get_mean_rewards_at_stages(result['rewards_history'])
        unblocked_summary_data.append({
            'condition': 'unblocked',
            'experiment': exp_idx + 1,
            't0': stage_rewards['t0'],
            't1': stage_rewards['t1'],
            't2': stage_rewards['t2'],
            't3': stage_rewards['t3'],
            'improvement': stage_rewards['t3'] - stage_rewards['t0']
        })
    
    # Combine data
    all_summary_data = blocked_summary_data + unblocked_summary_data
    df = pd.DataFrame(all_summary_data)
    
    # Print statistics
    print("\n" + "="*70)
    print("RESULTS SUMMARY")
    print("="*70)
    
    blocked_stats = aggregate_experiment_statistics(blocked_experiments)
    unblocked_stats = aggregate_experiment_statistics(unblocked_experiments)
    
    print("\nBLOCKED PATH:")
    print(f"  t0 (Initial): {blocked_stats['t0']['mean']:.4f} ± {blocked_stats['t0']['std']:.4f}")
    print(f"  t3 (Final):   {blocked_stats['t3']['mean']:.4f} ± {blocked_stats['t3']['std']:.4f}")
    print(f"  Improvement:  {blocked_stats['t3']['mean'] - blocked_stats['t0']['mean']:.4f}")
    
    print("\nUNBLOCKED PATH (Control):")
    print(f"  t0 (Initial): {unblocked_stats['t0']['mean']:.4f} ± {unblocked_stats['t0']['std']:.4f}")
    print(f"  t3 (Final):   {unblocked_stats['t3']['mean']:.4f} ± {unblocked_stats['t3']['std']:.4f}")
    print(f"  Improvement:  {unblocked_stats['t3']['mean'] - unblocked_stats['t0']['mean']:.4f}")
    
    print(f"\nDIFFERENCE (Unblocked - Blocked):")
    print(f"  Final reward: {unblocked_stats['t3']['mean'] - blocked_stats['t3']['mean']:.4f}")
    print(f"  Improvement:  {(unblocked_stats['t3']['mean'] - unblocked_stats['t0']['mean']) - (blocked_stats['t3']['mean'] - blocked_stats['t0']['mean']):.4f}")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Side-by-side comparison of final rewards
    ax1 = axes[0, 0]
    df.boxplot(column='t3', by='condition', ax=ax1)
    ax1.set_title('Final Reward: Blocked vs Unblocked Path')
    ax1.set_xlabel('Condition')
    ax1.set_ylabel('Final Reward (t3)')
    ax1.get_figure().suptitle('')
    
    # Plot 2: Reward progression comparison
    ax2 = axes[0, 1]
    for condition in ['blocked', 'unblocked']:
        cond_data = df[df['condition'] == condition]
        stages = ['t0', 't1', 't2', 't3']
        means = [cond_data[stage].mean() for stage in stages]
        stds = [cond_data[stage].std() for stage in stages]
        ax2.errorbar(stages, means, yerr=stds, marker='o', label=condition, capsize=5, linewidth=2)
    ax2.set_title('Reward Progression: Blocked vs Unblocked')
    ax2.set_xlabel('Stage')
    ax2.set_ylabel('Average Reward')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Improvement comparison
    ax3 = axes[1, 0]
    df.boxplot(column='improvement', by='condition', ax=ax3)
    ax3.set_title('Reward Improvement: Blocked vs Unblocked')
    ax3.set_xlabel('Condition')
    ax3.set_ylabel('Improvement (t3 - t0)')
    ax3.get_figure().suptitle('')
    ax3.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    
    # Plot 4: Violin plot for distribution comparison
    ax4 = axes[1, 1]
    blocked_t3 = df[df['condition'] == 'blocked']['t3']
    unblocked_t3 = df[df['condition'] == 'unblocked']['t3']
    ax4.violinplot([blocked_t3, unblocked_t3], positions=[1, 2], showmeans=True, showmedians=True)
    ax4.set_xticks([1, 2])
    ax4.set_xticklabels(['Blocked', 'Unblocked'])
    ax4.set_title('Final Reward Distribution')
    ax4.set_ylabel('Final Reward (t3)')
    ax4.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig('blocked_path_test.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Show example of blocked configuration
    print("\n--- Visualizing example blocked path configuration ---")
    example_obstacles = block_path_to_target(target_position, starting_position, num_blocking_obstacles)
    result = run_quantum_swarm_experiment(
        num_robots=num_robots,
        obstacle_positions=example_obstacles,
        target_position=target_position,
        starting_position=starting_position,
        delta_threshold=delta_threshold,
        verbose=True,
        plot=True,
        plot_all_steps=False
    )
    
    return {
        'blocked_results': blocked_experiments,
        'unblocked_results': unblocked_experiments,
        'dataframe': df,
        'blocked_stats': blocked_stats,
        'unblocked_stats': unblocked_stats
    }


def run_all_comprehensive_experiments():
    """
    Run all three comprehensive experiments and generate complete report.
    """
    print("\n" + "#"*70)
    print("#" + " "*68 + "#")
    print("#" + "  COMPREHENSIVE QUANTUM SWARM ROBOTICS COMPARISON STUDY".center(68) + "#")
    print("#" + " "*68 + "#")
    print("#"*70 + "\n")
    
    # Experiment 1: Swarm Size Comparison
    exp1_results = experiment_1_swarm_size_comparison(
        swarm_sizes=[2, 3, 5, 10],
        num_experiments_per_size=10,
        obstacle_positions=[(0.125, 0.125), (0.625, 0.625)],
        delta_threshold=0.8
    )
    
    # Export results
    exp1_results['dataframe'].to_csv('experiment1_swarm_size_comparison.csv', index=False)
    print("\n✓ Experiment 1 results saved to 'experiment1_swarm_size_comparison.csv'")
    
    # Experiment 2: Obstacle Density Comparison
    exp2_results = experiment_2_obstacle_density_comparison(
        obstacle_counts=[1, 2, 5, 10],
        num_experiments_per_density=10,
        num_robots=10,
        delta_threshold=0.8,
        use_random_obstacles=True
    )
    
    # Export results
    exp2_results['dataframe'].to_csv('experiment2_obstacle_density_comparison.csv', index=False)
    print("\n✓ Experiment 2 results saved to 'experiment2_obstacle_density_comparison.csv'")
    
    # Experiment 3: Blocked Path Test
    exp3_results = experiment_3_blocked_path_test(
        num_experiments=10,
        num_robots=10,
        num_blocking_obstacles=5,
        delta_threshold=0.8
    )
    
    # Export results
    exp3_results['dataframe'].to_csv('experiment3_blocked_path_test.csv', index=False)
    print("\n✓ Experiment 3 results saved to 'experiment3_blocked_path_test.csv'")
    
    # Generate summary report
    print("\n" + "="*70)
    print("COMPREHENSIVE SUMMARY REPORT")
    print("="*70)
    
    print("\n1. SWARM SIZE COMPARISON:")
    print("   Best performing swarm size:")
    best_size = exp1_results['dataframe'].groupby('swarm_size')['t3'].mean().idxmax()
    best_reward = exp1_results['dataframe'].groupby('swarm_size')['t3'].mean().max()
    print(f"   {best_size} robots with average final reward: {best_reward:.4f}")
    
    print("\n2. OBSTACLE DENSITY COMPARISON:")
    print("   Performance degradation with obstacles:")
    obs_summary = exp2_results['dataframe'].groupby('num_obstacles')['t3'].mean()
    print(f"   1 obstacle:  {obs_summary[1]:.4f}")
    print(f"   10 obstacles: {obs_summary[10]:.4f}")
    print(f"   Degradation: {obs_summary[1] - obs_summary[10]:.4f}")
    
    print("\n3. BLOCKED PATH TEST:")
    blocked_mean = exp3_results['dataframe'][exp3_results['dataframe']['condition'] == 'blocked']['t3'].mean()
    unblocked_mean = exp3_results['dataframe'][exp3_results['dataframe']['condition'] == 'unblocked']['t3'].mean()
    print(f"   Blocked path final reward:   {blocked_mean:.4f}")
    print(f"   Unblocked path final reward: {unblocked_mean:.4f}")
    print(f"   Impact of blocking:          {unblocked_mean - blocked_mean:.4f}")
    
    print("\n" + "="*70)
    print("All experiments completed successfully!")
    print("Figures and CSV files have been saved.")
    print("="*70 + "\n")
    
    return {
        'experiment1': exp1_results,
        'experiment2': exp2_results,
        'experiment3': exp3_results
    }


# Example usage
if __name__ == "__main__":
    # Run individual experiments or all at once
    
    # Option 1: Run all experiments
    all_results = run_all_comprehensive_experiments()
    
    # Option 2: Run individual experiments
    # exp1 = experiment_1_swarm_size_comparison()
    # exp2 = experiment_2_obstacle_density_comparison()
    # exp3 = experiment_3_blocked_path_test()
    
    # Option 3: Custom experiment combinations
    # custom_exp = experiment_1_swarm_size_comparison(
    #     swarm_sizes=[5, 15, 25],
    #     num_experiments_per_size=20
    # )
```
